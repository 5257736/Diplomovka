#### 1. Priraďte predpis funkcie ku grafu: 

$$\text{a) }z=2+x^2+y^2; \qquad \text{b) }z=2-x^2-y^2; \qquad \text{c) }z=2\left(x^2+y^2\right); \qquad \text{d) }z=2+2x-y; \qquad \text{e) }z=2.$$


Riešenie:   
<font color=blue> (a) (IV)  
(b) (II)  
(c) (I)  
(d) (V)  
(e) (III)  
</font>  

  

<ins> Vysvetlenie: </ins>   
Na základe grafov je možné určiť funkcie podľa ich charakteristík, pričom ich môžeme prepojiť s známymi 2D objektmi:

**(I)**: Graf zobrazuje **rotačný paraboloid** s vrcholom v bode $ z = 2 $ a symetricky otvorený nahor. Zodpovedá rovnici $ z = 2 + x^2 + y^2 $. Rezy rovinou kolmou na os $z$ sú kružnice. Rezy rovinou obsahujúcou os $z$ sú paraboly.

**(II)**: Tento graf zobrazuje **rotačný paraboloid** symetricky otvorený nadol, s vrcholom v $ z = 2 $. Zodpovedá rovnici $ z = 2 - x^2 - y^2 $. Rezy rovinou kolmou na os $z$ sú kružnice. Rezy rovinou obsahujúcou os $z$ sú paraboly otvorené nadol. 

**(III)**: Graf predstavuje **rovinu**, kde $ z $ má konštantnú hodnotu $ 2 $. Zodpovedá rovnici $z = 2 $. Rovina rovnobežná s rovinou $xy$. Všetky rezy sú priamky.  

**(IV)**: Tento graf je **rotačný paraboloid** s vrcholom v $ z = 2 $, ktorý je však násobený konštantou $ 2 $. Zodpovedá rovnici $ z = 2(x^2 + y^2) $.Rezy rovinou kolmou na os $z$ sú kružnice. Rezy rovinou obsahujúcou os $z$ sú paraboly, ktoré sú "užšie" ako v prípade (I). 

**(V)**: Rovinný graf má tvar **naklonenej roviny** s rovnicou závislou na $ x $ a $ y $. Zodpovedá rovnici $ z = 2 + 2x - y $. Všetky rezy sú priamky. Rezy rovinou kolmou na os $z$ sú priamky so smernicou závislou od sklonu roviny.

Tieto súvislosti pomáhajú vytvoriť intuíciu o tom, ako 2D objekty, ktoré poznáme (kružnice, paraboly, priamky), vytvárajú 3D povrchy pri "natiahnutí" do tretieho rozmeru. Názvy ako "paraboloid" alebo "hyperboloid" sú odvodené práve od tvarov, ktoré vznikajú pri rezoch týchto 3D objektov rovinami.

#### 2. Priraďte povrchy k príslušným vrstevnicovým grafom: 

   

Riešenie:  
<font color=blue> (a) (III)  
(b) (I)   
(c) (V)  
(d) (II)  
(e) (IV)  </font>    
  
  
<ins> Vysvetlenie: </ins>  
Vrstevnicové grafy (kontúrové diagramy) sú projekciou priestorových grafov na rovinu $xy$. Predstavte si, že na priestorový graf zasvietime svetlom zhora (smerom k osi $z$), a to, čo sa "odrazí" na rovinu $xy$, sú vrstevnice. Každá vrstevnica zodpovedá množine bodov, kde má funkcia rovnakú hodnotu $z=kon t$ . Tento koncept nám umožňuje vizualizovať správanie funkcie bez potreby trojrozmerného zobrazenia.

Súvis vrstevníc s priestorovými grafmi:
- Zhustené vrstevnice, znamenajú rýchlejší rast alebo pokles funkčných hodnôt v danom smere.
- Riedke vrstevnice, znamenajú pomalší rast alebo pokles funkčných hodnôt.
- Znamienka funkčných hodnôt, vrstevnice s kladnými hodnotami $z$ sú odlíšené od tých so zápornými hodnotami, čo je viditeľné z popisov hodnôt na grafe.

Prepojenie jednotlivých grafov:  
**(a) (III)**:

Priestorový graf : Rovina $z=2$, ktorá je rovnobežná s rovinou $xy$.    
Vrstevnicový graf: Sú to sústredné kružnice so stredom v bode $(0,0)$, pretože všetky body majú rovnakú hodnotu $z=2$.  

**(b) (I)**:  

Priestorový graf: Rovina $z=2+y$, ktorá stúpa pozdĺž osi $y$.     
Vrstevnicový graf: Paralelné čiary v rovine $xy$, kde každá čiara zodpovedá konštantnej hodnote $z$.   

**(c) (V)**:   

Priestorový graf: Naklonená rovina $z=2+2x−y$.    
Vrstevnicový graf: Paralelné čiary, ktoré sú šikmo orientované, pričom ich sklon závisí od koeficientov pri $x$ a $y$.      

**(d) (II)**:  

Priestorový graf: Paraboloid otvorený nadol, $z=2−x^2-y^2$.   
Vrstevnicový graf: Sústredné kružnice so stredom v bode $(0,0)$, pričom hodnoty vrstiev klesajú smerom von.
 
**(e) (IV)**:   

Priestorový graf: Paraboloid otvorený nahor, $z=2(x^2 +y^2)$.  
Vrstevnicový graf: Sústredné kružnice so stredom v bode $(0,0)$, pričom hodnoty vrstiev rastú smerom von.

Týmto spôsobom môžeme intuitívne pochopiť prepojenie medzi vrstevnicovými diagramami a priestorovými povrchmi. Tento prístup zdôrazňuje dôležitosť hustoty vrstevníc a ich tvaru pri interpretácii správania funkcie v priestore.

#### 3. V trojrozmernom priestore je množina bodov spĺňajúca rovnicu $y=1$ priamka rovnobežná s osou $x$. 

a) Áno     
<font color=blue> b) Nie </font>
  
<ins> Vysvetlenie: </ins>    
V trojrozmernom priestore množina bodov spĺňajúca rovnicu $y = 1$ predstavuje rovinu, nie priamku. Rovnica $y = 1$ v trojrozmernom priestore teda definuje nekonečnú rovinu, ktorá je rovnobežná s rovinou $xz$, vo výške $y=1$ resp. pretína os $y$ v bode $1$ a prechádza bodmi, kde sú všetky hodnoty $x$ a $z$ možné, zatiaľ čo hodnota $y$ je vždy pevne $1$. 

Je dôležité zdôrazniť, že priamka v 3D priestore nemá všeobecnú rovnicu v tvare jednej rovnice. Dôvodom je, že v 3D priestore existuje nekonečne veľa normálových vektorov kolmých na danú priamku. Na jednoznačné určenie priamky v 3D priestore je potrebné použiť buď:
1. Dve rovnice rovín, ktorých priesečnica definuje priamku.
2. Parametrické vyjadrenie priamky.
3. Bod a smerový vektor priamky.

Toto je zásadný rozdiel oproti 2D priestoru, kde jedna rovnica môže jednoznačne definovať priamku.

#### 4. Rovina $xy$ má rovnicu $xy=0$. 

a) Áno     
<font color=blue> b) Nie </font>    
  
<ins> Vysvetlenie: </ins>    
Rovnica $xy = 0$ definuje množinu bodov, kde aspoň jedna zo súradníc $x$ alebo $y$ musí byť nulová, čo vytvára dve na seba kolmé roviny prechádzajúce počiatkom súradnicovej sústavy, nie rovinu $xy$.   
Rovina $xy$ má v priestore rovnicu $z = 0$, pretože:
- obsahuje všetky body, ktorých z-ová súradnica je $0$
- $x$ a $y$ súradnice môžu nadobúdať ľubovoľné reálne hodnoty
- je kolmá na os $z$

#### 5. Ak $f(x,y)$ je funkcia dvoch premenných definovaná pre všetky $x$ a $y$, potom $f(10,y)$ je funkcia jednej premennej. 

<font color=blue> a) Áno    </font>   
b) Nie   
  
<ins> Vysvetlenie: </ins>    
Keď máme funkciu dvoch premenných $f(x,y)$, každej dvojici $(x,y)$ je jednoznačne priradená hodnota $z = f(x,y)$. Ak fixujeme hodnotu jednej premennej (v tomto prípade $x = 10$), funkčná závislosť sa neporuší. Zvyšnej premennej $y$ bude stále jednoznačne priradená jedna hodnota, čím vznikne funkcia jednej premennej.

Napríklad:  
$f(x,y) = x^2 + y^2$  
Po dosadení $x = 10$ dostaneme:
$$f(10,y) = 100 + y^2$$
 
Táto nová funkcia už závisí len od premennej $y$. Pre každú hodnotu $y$ existuje práve jedna hodnota $f(10,y)$, čo je definícia funkcie jednej premennej. Napríklad:

Ak $y = 0$, potom $f(10,0) = 100$   
Ak $y = 1$, potom $f(10,1) = 101$   
Ak $y = 2$, potom $f(10,2) = 104$  

Tento princíp platí pre všetky funkcie dvoch premenných - fixovaním jednej premennej vždy získame funkciu jednej premennej.

#### 6. Objem $V$ krabice s výškou $h$ a so stranou $s$ štvorcovej podstavy je funkciou premenných $h$ a $s$. 

<font color=blue> a) Áno    </font>   
b) Nie   
  
<ins> Vysvetlenie: </ins>    
Objem krabice so štvorcovou podstavou sa vypočíta ako $V = s^2 \cdot h$, $s$ je dĺžka strany štvorcovej podstavy a $h$ je výška krabice.
Pre každú dvojicu hodnôt $(h,s)$ existuje práve jedna hodnota objemu $V$.
Zmena ktorejkoľvek z premenných ($h$ alebo $s$) ovplyvní výslednú hodnotu objemu $V$.  
Objem $V$ je teda závislý od **dvoch nezávislých** premenných $h$ a $s$, čo z neho robí funkciu dvoch premenných $V(h,s)$.

#### 7. Ak $f(x)$ a $g(y)$ sú funkcie jednej premennej, potom ich súčin $f(x)\cdot g(y)$ je funkciou dvoch premenných.

<font color=blue> a) Áno    </font>   
b) Nie   
  
<ins> Vysvetlenie: </ins>  
Funkcia $f(x)$ závisí len od premennej $x$. Funkcia $g(y)$ závisí len od premennej $y$. Ich súčin $h(x,y) = f(x)·g(y)$ závisí od oboch premenných $x$ aj $y$.

Napríklad:
Ak $f(x) = x$ a $g(y) = y^2$. Potom $h(x,y) = f(x)\cdot g(y) = x \cdot y^2$. Hodnota $h(x,y)$ sa mení so zmenou $x$ aj so zmenou $y$. Pre každú dvojicu hodnôt $(x,y)$ existuje práve jedna hodnota $h(x,y)$.

#### 8. Priamka rovnobežná s osou $z$ môže pretínať graf funkcie $f(x,y)$ najviac raz. 

<font color=blue> a) Áno    </font>   
b) Nie    
  
<ins> Vysvetlenie: </ins>    
Priamka rovnobežná s osou  $ z  $ má pevné  $ x  $ a  $ y  $ (napríklad  $ x = c_1  $,  $ y = c_2  $). Keďže graf funkcie  $ f(x, y)  $ priraďuje každej dvojici  $(x, y)  $ práve jednu hodnotu  $ z  $, priamka môže pretínať graf funkcie najviac v **jednom bode**.

#### 9. Graf funkcie $f(x,y)=x^2+y^2$ je kruh. 

a) Áno     
<font color=blue> b) Nie </font>    
  
<ins> Vysvetlenie: </ins>  
Graf funkcie $f(x,y)=x^2+y^2$ nie je kruh, pretože zobrazuje všetky trojice $(x,y,z)$, kde $z=x^2+y^2$. Ide o trojrozmerný povrch nazývaný **paraboloid**.  
Rovnica kruhu je dvojrozmerná, napríklad $x^2+y^2=r^2$, kde $r$ je polomer. Táto rovnica neobsahuje $z$, takže opisuje útvar iba v rovine $xy$.

#### 10. Vrstevnicový diagram funkcie $z=f(x,y)$ je plocha v priestore $xyz$.

a) Áno     
<font color=blue> b) Nie </font>    
  
<ins> Vysvetlenie: </ins>  
Vrstevnicový diagram (**Definícia 4**) funkcie $z = f(x,y)$ nie je plocha v priestore $xyz$, ale je to **rovinný obrázok**   v rovine $xy$  (pozostávajúci z kriviek v tejto rovine (2D)), ktorý sa používa na vizualizáciu správania funkcie dvoch premenných.  

Vrstevnice sú množiny bodov v rovine $xy$, pričom:
- Každá vrstevnica predstavuje množinu bodov $(x,y)$, pre ktoré funkcia $f(x,y)$ nadobúda konštantnú hodnotu.
- Rôzne vrstevnice zodpovedajú rôznym hodnotám funkcie $f(x,y)$.

Inými slovami, na každej jednotlivej vrstevnici je hodnota funkcie $f(x,y)$ konštantná, ale medzi rôznymi vrstevnicami sa táto hodnota mení. Tento koncept umožňuje vizualizovať, ako sa mení hodnota funkcie dvoch premenných v rovine $xy$ bez potreby tretieho rozmeru.

#### 11. Ak sa vrstevnice funkcie $f(x,y)$ v určitom smere približujú viac k sebe, potom funkcia v tomto smere rastie.

a) Áno     
<font color=blue> b) Nie </font>    
  
<ins> Vysvetlenie: </ins>   
Ak sa vrstevnice funkcie $f(x,y)$ v určitom smere približujú k sebe, znamená to, že sa funkcia v tomto smere **mení rýchlejšie/prudšie** (**Príklad 10**) (má väčší gradient), ale nehovorí to nič o tom, či funkcia rastie alebo klesá - o tom rozhodujú funkčné hodnoty na po sebe nasledujúcich vrstevniciach.

#### 2.  Transponovaná matica k matici   $\boldsymbol{\mathrm{D}}=\left(\begin{array}{rrrr}
2 & -3 & 9 & 7
\end{array}\right)$
je:  

a) $\left(\begin{array}{rrrr}
2 & -3 & 9 & 7
\end{array}\right)$

b) $\left(\begin{array}{rrrr}
7 & 9 & -3 & 2
\end{array}\right)$

<font color=blue> c) $\left(\begin{array}{r}
2 \\
-3 \\
9 \\
7
\end{array}\right)$   </font>

d) $\left(\begin{array}{r}
7 \\
9 \\
-3 \\
2
\end{array}\right)$

E) Žiadna z odpovedí A), B), C), D) nie je správna.  
  
<ins> Vysvetlenie: </ins>     
**_Transponovaná matica_** k matici $\boldsymbol{\mathrm{A}}$ typu $m\times n$ je matica $\boldsymbol{\mathrm{A}}^{T}$ resp. $\boldsymbol{\mathrm{A}}^{'}$ typu $n\times m$ s prvkami $a_{ij}^T$ taká, že $a_{ij}^T=a_{ji}$ pre každé $i\in\{1,\ldots,m\}$, $j\in\{1,\ldots,n\}$.  

V našom prípade ide o riadkový vektor, z ktorého spravíme stĺpcový vektor.

#### 3. Majme dané matice a s nasledujúcimi rozmermi: $\boldsymbol{\mathrm{A}}$: $2\times3$, $\boldsymbol{\mathrm{B}}$: $3\times5$, $\boldsymbol{\mathrm{C}}$: $5\times1$, $\boldsymbol{\mathrm{D}}$: $1\times5$. Ktoré z nasledujúcich súčinov sú definované?

<font color=blue>a) AB  </font>      
b) BA   
c) CA  
<font color=blue> d) CD </font>   
<font color=blue> e) DC  </font>  
f) AC  
<font color=blue> g) (AB)C  </font>  
  
<ins> Vysvetlenie: </ins>   
**Definícia 1** hovorí, že zložené číslo $a\not\in \{−1,0,1\}$ je také číslo, ktoré má netriviálne delitele. Okrem deliteľov $−1, 1, −a, a$ má aj iné delitele. Takže má viac ako dvoch kladných deliteľov. **Tvrdenie 1** nám hovorí aj to, že musí existovať $b,c\in \mathbb{Z}$ také, že $a=bc$ a $1 < |b|$, $|c| < |a|$.

#### 10. Platí?  $$\mathrm{det}\left(\boldsymbol{\mathrm{A}}+\boldsymbol{\mathrm{B}}\right)=\mathrm{\det}\left(\boldsymbol{\mathrm{A}}\right)+\mathrm{\det}\left(\boldsymbol{\mathrm{B}}\right).$$

a) Áno    
<font color=blue> b) Nie </font>    
 
  
<ins> Vysvetlenie: </ins>  
V **Poznámke 9** máme spomenuté, že to neplatí vo všeobecnosti.

Napr. máme matice $2\times2$
matica $A$ je jednotková a matica $B$ má po diagonále čísla $5$ a mimo diagonály $0$. Potom $\mathrm{\det}\left(\boldsymbol{\mathrm{A}}\right)=1$ a $\mathrm{\det}\left(\boldsymbol{\mathrm{B}}\right)=25$. Matica $A+B$ je taká matica, ktorá má po diagonále $6$ a mimo nej nuly. Teda $\mathrm{det}\left(\boldsymbol{\mathrm{A}}+\boldsymbol{\mathrm{B}}\right)=36$, čo nie je rovné $\mathrm{\det}\left(\boldsymbol{\mathrm{A}}\right)+\mathrm{\det}\left(\boldsymbol{\mathrm{B}}\right)=1+25=26$.

#### 11. Ak $\boldsymbol{\mathrm{A}}$ je $3\times3$ matica, potom

$$\mathrm{\det}\left(5\cdot\boldsymbol{\mathrm{A}}\right)=5\cdot\mathrm{\det}\left(\boldsymbol{\mathrm{A}}\right).$$

a) Áno    
<font color=blue> b) Nie </font>    
 
  
<ins> Vysvetlenie: </ins>  
Neplatí to vo všeobecnosti.

Kontrapríklad:
Zoberieme si jednotkovú maticu $A$, $2\times2$. Vieme, že $\mathrm{\det}\left(\boldsymbol{\mathrm{A}}\right)=1$, čiže potom matica $5\cdot A$ je taká matica, ktorá má po diagonále číslo $5$ a mimo nej nuly. Potom $\mathrm{\det}\left(\boldsymbol{\mathrm{5\cdot A}}\right)=25$, ale $5\cdot \mathrm{\det}\left(\boldsymbol{\mathrm{A}}\right)=5\cdot 1=5$ vidíme, že $5$ nie je rovné $25$.

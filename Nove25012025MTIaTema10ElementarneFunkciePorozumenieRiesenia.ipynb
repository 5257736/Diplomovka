#### 2.  Je nasledujúce tvrdenie pravdivé?  
#### Funkcia $f(x)=e^{2x}/\left(2e^{5x}\right)$ je exponenciálna.

<font color=blue> a) Áno   </font>    
b) Nie  
  
<ins> Vysvetlenie: </ins>     
Je potrebné zjednodušiť predpis funkcie, aby sme vedeli rozhodnúť o tom, že či je zadaná funkcia exponenciálna alebo nie.

Zjednodušenie predpisu funkcie:
$$\frac{e^{2x}}{2e^{5x}}=\frac{1}{2}\cdot e^{2x}.e^{-5x}=\frac{1}{2}.e^{2x-5x}=\frac{1}{2}.e^{-3x}$$

Keďže funkcia $e^{-3x}$ je exponenciálna funkcia, $f(x)=\frac{1}{2}.e^{-3x}$ je taktiež exponenciálna funkcia, pretože je to len násobok exponenciálnej funkcie. Takže aj pôvodný predpis funkcie predstavuje exponenciálnu funkciu, pretože sme použili ekvivalentné úpravy.

#### 3. Je nasledujúce tvrdenie pravdivé?  
#### Exponenciálna funkcia môže byť klesajúca.

<font color=blue> a) Áno    </font>   
b) Nie  
  
<ins> Vysvetlenie: </ins>   
Exponenciálna funkcia má tvar $f(x)=a⋅b^x$, kde $a$ je konštanta a $b$ je základ exponenciálnej funkcie.

Pre rastúcu exponenciálnu funkciu: Ak $b>1$, funkcia je rastúca, pretože hodnota funkcie sa zvyšuje, keď $x$ rastie.

Pre klesajúcu exponenciálnu funkciu: Ak $0<b<1$, funkcia je klesajúca. V tomto prípade, keď $x$ rastie, hodnota funkcie sa znižuje, pretože základ $b$ je menší ako $1$.

Pre konkrétny príklad, ak máme funkciu $f(x)=2⋅(0.5)^x$, táto funkcia je klesajúca, pretože základ $0.5$ je menší ako $1$.

Takže áno, exponenciálna funkcia môže byť klesajúca, ak základ $b$ je medzi $0$ a $1$.

#### 4. Je nasledujúce tvrdenie pravdivé?   
#### Funkcia $-\log|x|$ je nepárna.

a) Áno      
<font color=blue> b) Nie  </font>   
  
<ins> Vysvetlenie: </ins>    
Musíme skontrolovať, či platí podmienka pre nepárne funkcie:  
$$f(−x)=−f(x)$$  
Pre našu funkciu $f(x)=−\log⁡∣x∣$, skúsme vypočítať $f(−x)$:  
$$f(−x)=−\log⁡∣−x∣$$    
Vzhľadom na vlastnosti absolútnej hodnoty, $∣−x∣=∣x∣$, teda:    
$$f(−x)=−\log⁡∣x∣$$   
Teraz porovnáme $f(−x)$ s $−f(x)$:   
$$−f(x)=−(−\log⁡∣x∣)=\log⁡∣x∣$$   
Vidíme, že:   
$$f(−x)=−\log⁡∣x∣≠\log⁡∣x∣=−f(x)$$   

Funkcia $−\log⁡∣x∣$ nie je nepárna, pretože $f(−x)≠−f(x)$.

```python
x = var('x')
f(x) = - log(abs(x))
show(LatexExpr('f(x) = ')+latex(f(x)))
plot(f(x), (x, -4, 4), axes_labels=['$x$', '$y$'], axes_labels_size = 1.2, color='blue')
```

#### 5. Je nasledujúce tvrdenie pravdivé? 
#### $\forall x >0:$ $\ln x > \log x$.

a) Áno   
<font color=blue> b) Nie  </font>   
  
<ins> Vysvetlenie: </ins>     
Je potrebné si vykresliť obe funkcie do jedného grafu, kde uvidíme, že pre $0<x<1$ je $\ln x <\log x$ a pre $x>1$ platí $\ln x>\log x$.  

Dôležité je premeniť si to na rovnaké logaritmy. Zoberieme si $\ln x$ a ten zmeníme nasledovne:  
$$\ln x=\log_{e}x=\frac{\log x}{\log e}$$
Táto rovnosť vyplýva zo vzťahu medzi prirodzeným logaritmom (ln) a logaritmom so základom 10 (log). Prirodzený logaritmus je logaritmus so základom $e$, takže $\ln x = \log_{e} x$. Potom použijeme vzorec pre zmenu základu logaritmu: $\log_{a} x = \frac{\log x}{\log a}$, kde v našom prípade $a = e$.  

Teraz to porovnáme, že či platí:
$$\ln x>\log x$$
$$\frac{\log x}{\log e}>\log x$$
Vyčíslime $\log e=0,4343$.

$$\frac{\log x}{0,4343}>\log x$$

Napr.
$$x=0,1$$  

$$\frac{\log 0,1}{0,4343}>\log 0,1$$   

$$\frac{-1}{0,4343}> -1$$  

$$-2,3> -1  $$ Nerovnica **NEPLATÍ**

$$x=100$$  

$$\frac{\log 100}{0,4343}>\log 100$$  

$$\frac{2}{0,4343}> 2$$  

$$4,6> 2  $$Nerovnica **PLATÍ**

Nerovnica platí pre všetky $x>1$ a nie pre všetky $x>0$. 

```python
x = var('x')
f(x) = log(x,10)
g(x) = ln(x)
show(LatexExpr('f(x) = \\log x')) 
show(LatexExpr('g(x) = \\ln x'))
kwds = {'fontsize':12}
plot(f(x), (x, 0.1, 4), axes_labels=['$x$', '$y$'], axes_labels_size = 1.2, color='blue') + plot(g(x), (x, 0.1, 4), axes_labels=['$x$', '$y$'], axes_labels_size = 1.2, color='red') + text("$y=\\ln\\, x$", (3, 1.4), color='red', **kwds) + text("$y=\\log\\, x$", (3, 0.7), color='blue', **kwds)
```

#### 6. Je nasledujúce tvrdenie pravdivé? 
#### Graf funkcie $g(x)=\log(x-1)$ pretína os $x$ v bode $1$.

a) Áno  
<font color=blue> b) Nie  </font>   
  
<ins> Vysvetlenie: </ins>     
Aby sme zistili, či graf funkcie $g(x)=\log⁡(x−1)$ pretína os $x$ v bode $x=1$, musíme určiť, kedy hodnota funkcie $g(x)$ je nula. To znamená, že hľadáme hodnotu $x$, pre ktorú platí: $\log⁡(x−1)=0$  

Pri logaritmických funkciách platí, že: $\log⁡_b(y)=0$ práve vtedy, keď $y=1$  

V tomto prípade: $x−1=1$   

Teda:  $x=2$  

To znamená, že $g(x)$ pretína os $x$ v bode $x=2$, nie v bode $x=1$.

```python
x = var('x')
f(x) = log(x-1)
show(LatexExpr('f(x) = \\log(x-1)'))
plot(f(x), (x, 1.1, 4), axes_labels=['$x$', '$y$'], axes_labels_size = 1.2, color='blue') + line(([1,1.5],[1,-2.5]), color='orange', linestyle='dashed')
```

#### 7. Je nasledujúce tvrdenie pravdivé? 
#### Inverzná funkcia k funkcii $y=\log\,x$ je $y=\frac{1}{\log x}$.

a) Áno   
<font color=blue> b) Nie  </font>    
  
<ins> Vysvetlenie: </ins>   
Aby sme našli inverznú funkciu k $y=\log ⁡x$, musíme vyjadriť $x$ ako funkciu $y$.  

Začnime s pôvodnou funkciou: $y=\log⁡x$   
Premeníme to na exponenciálny tvar: $x=10^y$  
Tento krok je kľúčový - využívame tu definíciu logaritmu, kde $\log x = y$ je ekvivalentné s $10^y = x$.  
Takže inverzná funkcia k $y=\log⁡ x$ je: $y=10^x$  
Inverzná funkcia k $y=\log⁡ x$ je teda $y=10^x$ a nie $y=\frac{1}{\log(x)}$.   

Je dôležité poznamenať, že zápis $f^{-1}$ pre inverznú funkciu nie je to isté ako $\frac{1}{f}$. Ide len o konvenciu v označovaní, ktorá môže byť mätúca. Inverzná funkcia "obracia" pôvodnú funkciu, zatiaľ čo $\frac{1}{f}$ je len recipročná hodnota funkcie.

#### 8. Je nasledujúce tvrdenie pravdivé? 
#### Funkcie $f(x)=3\cos x$ a $g(x)=\cos3x$ nemajú rovnakú periódu.

<font color=blue> a) Áno  </font>       
 b) Nie  

<ins> Vysvetlenie: </ins>  
Musíme určiť periódu každej funkcie samostatne.

Funkcia $f(x)=3\cos⁡(x)$
Funkcia $f(x)$ je len násobok základnej kosínusovej funkcie $\cos(x)$. Násobenie amplitúdy nemá vplyv na periódu funkcie. Perióda základnej funkcie $\cos⁡(x)$ je $2\pi$. Preto aj funkcia $f(x)$ má periódu $2\pi$.

Funkcia $g(x)=\cos⁡(3x)$
Funkcia $g(x)$ je kosínusová funkcia so zmenou v argumente. Ak je argument funkcie násobený číslom $k$, perióda základnej funkcie $\cos⁡(x)$ sa delí týmto číslom $k$. Preto perióda funkcie $\cos(3x)$ je $2\frac{\pi}{3}$.  

Perióda funkcie $f(x)=3\cos⁡(x)$ je $2\pi$.  
Perióda funkcie $g(x)=\cos⁡(3x)$ je $2\frac{\pi}{3}$​.  

Funkcie nemajú rovnakú periódu. Násobenie funkcie číslom 3 v prípade $f(x)=3\cos x$ ovplyvňuje obor hodnôt funkcie, nie jej periódu. Obor hodnôt funkcie $f(x)$ je interval $[-3,3]$, zatiaľ čo obor hodnôt funkcie $g(x)=\cos 3x$ zostáva $[-1,1]$.

```python
x = var('x')
f(x) = 3*cos(x) 
g(x) = cos(3*x)
xt = [k * pi/2 for k in range(-5, 6)] 
yt = [-3,-2,-1,0,1,2,3]
xt_lbl = ['$' + latex(t) + '$' for t in xt]
yt_lbl = ['$' + latex(t) + '$' for t in yt]
plot(f(x), (x, -8, 8), axes_labels=['$x$', '$y$'], ticks=[xt, yt], tick_formatter=[xt_lbl, yt_lbl], axes_labels_size = 1.2, color='blue') + plot(g(x), (x, -8, 8), axes_labels=['$x$', '$y$'], axes_labels_size = 1.2, color='red') + text("$f(x)=3\\cos (x)$", (10,2), color='blue') + text("$g(x)=\\cos\\, 3x$", (10,1.5), color='red') 
```

#### 9. Je nasledujúce tvrdenie pravdivé?  
#### Funkcia $|\sin\,x|$ je párna.

<font color=blue> a) Áno   </font>     
b) Nie  
  
<ins> Vysvetlenie: </ins>  
Musí byť splnená podmienka pre párnu funkciu $f(−x)=f(x)$ pre všetky $x$ v jej definičnom obore.

Overenie:  
$$f(x)=∣sinx∣$$

$$f(−x)=∣sin(−x)∣$$

Je potrebné využiť nepárnosť sínusovej funkcie: 
$\sin(−x)=−\sin(x)$  

Preto platí: $f(−x)=∣\sin(−x)∣=∣-\sin(x)∣=∣\sin(x)∣$  

Vidíme, že $f(−x)=f(x)$, $∣\sin(x)∣=∣\sin(x)∣$ pre všetky $x$.

To znamená, že funkcia $∣\sin x∣$ je **párna**.

```python
x = var('x')
f(x) = abs(sin(x))
show(LatexExpr('f(x) = ') + latex(f(x)))
plot(f(x), (x, -8, 8), axes_labels=['$x$', '$y$'], axes_labels_size = 1.2, color='blue')
```

#### 10. Je nasledujúce tvrdenie pravdivé? 
#### Zloženie dvoch kvadratických funkcií je opäť kvadratická funkcia.

a) Áno      
<font color=blue> b) Nie  </font>
 
  
<ins> Vysvetlenie: </ins>  
Pre lepšie vysvetlenie si zloženie dvoch kvadratických rozpíšeme po krokoch:  

1. Máme dve kvadratické funkcie $f(x)=ax^2+bx+c$ a $g(x)=dx^2+ex+f$, a chceme zložiť $f(x)$ s $g(x)$ (tj. $f(g(x))$), musíme nahradiť $x$ vo funkcii $f$ výrazom $g(x)$:  

$$f(g(x))=a(g(x))^2+b(g(x))+c$$

2. Teraz vypočítajme $(g(x))^2$:  
$$(g(x))^2=(dx^2+ex+f)^2$$

Po umocnení dostaneme $$(dx^2+ex+f)^2=d^2x^4+(2de+e^2+2df)x^2+2efx+f^2$$

3. Substituujeme do funkcie $f$, upravíme a dostaneme:  
$$f(g(x))=ad^2x^4+(a(2de+e^2+2df)+bd)x^2+(2aef+be)x+(af^2+bf+c)$$

Ako vidíme, výsledná funkcia obsahuje člen $x^4$ (ak $a\ne 0$), čo znamená, že výsledná funkcia $f(g(x))$ je polynóm štvrtého stupňa, nie kvadratická funkcia.

Preto zloženie dvoch kvadratických funkcií nemusí byť opäť kvadratická funkcia (ak $a\ne0$).   
Pre $a=0$ zložením dvoch kvadratických funkcií dostaneme kvadratickú funkciu. 

Napr.  
Vezmime si jednoduché kvadratické funkcie $f(x)=x^2$ a $g(x)=x^2+1$.   
Ich zložením dostaneme:  
$f(g(x)) = f(x^2+1) = (x^2+1)^2 = x^4 + 2x^2 + 1$
Vidíme, že výsledkom je polynóm 4. stupňa, nie kvadratická funkcia.

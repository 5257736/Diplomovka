Následne definujte nové funkcie podľa Definície 3, t.j. vytvorte funkcie $f+g$, $g+f$, $f-g$, $g-f$, $f\cdot g$, $g\cdot f$, $\dfrac{f}{g}$, $\dfrac{g}{f}$, $f+c$, $c+f$, $f-c$, $c-f$, $f\cdot c$, $c\cdot f$, $\dfrac{f}{c}$, $\dfrac{c}{f}$. Predpokladajme, že máme dve funkcie:  
- $ f(x) = 2x + 1 $
- $ g(x) = x^2 - 4 $
- Konštantná funkcia $ c(x) = 5$

**Vytvorenie nových funkcií**

Podľa zadania môžeme vytvoriť nasledujúce nové funkcie:

1. $ f + g = (2x + 1) + (x^2 - 4) = x^2 + 2x - 3 $
2. $ g + f = (x^2 - 4) + (2x + 1) = x^2 + 2x - 3 $ (rovnaké ako $ f + g $)
3. $ f - g = (2x + 1) - (x^2 - 4) = -x^2 + 2x + 5 $
4. $ g - f = (x^2 - 4) - (2x + 1) = x^2 - 2x - 5 $
5. $ f \cdot g = (2x + 1) \cdot (x^2 - 4) = 2x^3 - 8x + x^2 - 4 $
6. $ g \cdot f = (x^2 - 4) \cdot (2x + 1) = 2x^3 - 8x + x^2 - 4$ (rovnaké ako $ f \cdot g $)
7. $ \frac{f}{g} = \frac{2x + 1}{x^2 - 4} $
8. $ \frac{g}{f} = \frac{x^2 - 4}{2x + 1} $
9. $ f + c = (2x + 1) + 5 = 2x + 6 $
10. $ c + f = 5 + (2x + 1) = 2x + 6 $ (rovnaké ako $ f + c $)
11. $ f - c = (2x + 1) - 5 = 2x - 4 $
12. $ c - f = 5 - (2x + 1) = -2x + 4 $
13. $ f \cdot c = (2x + 1) \cdot 5 = 10x + 5 $
14. $ c \cdot f = 5 \cdot (2x + 1) = 10x + 5 $ (rovnaké ako $ f \cdot c $)
15. $ \frac{f}{c} = \frac{2x + 1}{5} $
16. $ \frac{c}{f} = \frac{5}{2x + 1} $

**Zaujímavé javy**
- **Komutativita sčítania a násobenia**: Funkcie $ f + g $ a $ g + f $ sú rovnaké, rovnako ako $ f \cdot g $ a $ g \cdot f $, čo ukazuje komutativitu týchto operácií.
- **Posunutie grafov**: Operácie ako $ f + c $ a $ f - c $ posúvajú grafy funkcie $ f $ vertikálne hore alebo dole.
- **Zmena tvaru**: Násobenie alebo delenie funkcií môže zmeniť tvar grafu.


```python
# Definovanie funkcií
var('x')
f = 2*x + 1
g = x^2 - 4
c = 5

# Definovanie nových funkcií
f_plus_g = f + g
g_plus_f = g + f
f_minus_g = f - g
g_minus_f = g - f
f_times_g = f * g
g_times_f = g * f
f_div_g = f / g
g_div_f = g / f
f_plus_c = f + c
c_plus_f = c + f
f_minus_c = f - c
c_minus_f = c - f
f_times_c = f * c
c_times_f = c * f
f_div_c = f / c
c_div_f = c / f

# Vytvorenie grafov
P1 = plot(f, (x, -5, 5), color='red', legend_label='f(x) = 2x + 1')
P2 = plot(g, (x, -5, 5), color='green', legend_label='g(x) = x^2 - 4')
P3 = plot(f_plus_g, (x, -5, 5), color='blue', legend_label='f + g')
P4 = plot(f_minus_g, (x, -5, 5), color='purple', legend_label='f - g')
P5 = plot(f_times_g, (x, -5, 5), color='orange', legend_label='f * g')
P6 = plot(f_div_g, (x, -5, 5), color='black', legend_label='f / g')
P7 = plot(f_plus_c, (x, -5, 5), color='cyan', legend_label='f + c')
P8 = plot(f_minus_c, (x, -5, 5), color='magenta', legend_label='f - c')
```

**4. Uvažujte funkciu $f(x)=\dfrac{(x-1)(x-2)}{x-2}$ a $g(x)=x-1$. Na akej množine (pre aké $x$) sa tieto funkcie rovnajú?**  


<ins> Riešenie: <ins>  

Máme dve funkcie:  $f(x) = \frac{(x-1)(x-2)}{x-2}, \quad g(x) = x - 1$

- **Určenie definičných oborov**  
    - Funkcia $ f(x) $ obsahuje zlomok s menovateľom $ x - 2 $, ktorý nemôže byť nulový, teda musí platiť: $x - 2 \neq 0 \Rightarrow x \neq 2$. Definičný obor funkcie $ f(x) $ je teda $D_f = \mathbb{R} \setminus \{2\}$.
    - Funkcia $g(x) = x - 1$ je definovaná pre všetky reálne čísla, teda: $D_g = \mathbb{R}$

- **Úprava funkcie $ f(x) $**  
    Zjednodušme výraz pre $ f(x) $, pokiaľ to ide: $f(x) = \frac{(x-1)(x-2)}{x-2}$
    Pre $ x \neq 2 $ môžeme výraz $ x-2 $ v čitateli a menovateli vykrátiť: $f(x) = x - 1, \quad \text{pre všetky } x \neq 2.$
    Vidíme, že $ f(x) $ a $g(x) $ majú rovnaký predpis, ale $ f(x)$ nie je definovaná v bode $ x = 2$.
  
- Funkcie $ f(x) $ a $ g(x) $ majú rovnaký predpis na množine $ \mathbb{R} \setminus \{2\} $. Preto sa rovnajú na množine: $$\mathbb{R} \setminus \{2\} = (-\infty, 2) \cup (2, \infty)$$

**Záver:** Funkcie $ f(x) $ a $ g(x) $ sú rovnaké pre všetky $ x \neq 2 $, ale nie sú rovnaké na celej $ \mathbb{R} $, pretože $ f(x) $ nie je definovaná v bode $ x=2 $, zatiaľ čo $ g(x)$ áno.

Tento interaktívny kód vám umožňuje vykonávať rôzne operácie s funkciami. Na príklade funkcií $ f(x) = 2x + 1 $ a $ g(x) = x^2 - 4 $ sa dajú demonštrovať rôzne operácie.

Kód zobrazuje aj grafy týchto funkcií v zadanom intervale. Môžete si zvoliť akúkoľvek operáciu a sledovať, ako sa funkcie menia a vizualizujú v grafickej podobe. Tento interaktívny nástroj vám umožňuje efektívne vidieť výsledky rôznych operácií s funkciami a lepšie pochopiť, ako sa tieto operácie vykonávajú.

```python
x = var('x')
@interact
def _(f=2*x+1, g=x^2-4, xrange=input_box((0,1)), yrange='auto', a=1,
      action=selector(['f', 'df/dx', 'int f', 'num f', 'den f', '1/f', 'finv',
                       'f+a', 'f-a', 'f*a', 'f/a', 'f^a', 'f(x+a)', 'f(x*a)',
                       'f+g', 'f-g', 'f*g', 'f/g', 'f(g)'],
             width=15, nrows=5, label="h = "),
      do_plot = ("Draw Plots", True)):
    try:
        f = SR(f); g = SR(g); a = SR(a)
    except TypeError as msg:
        print(msg[-200:])
        print("Unable to make sense of f,g, or a as symbolic expressions.")
        return
    if not (isinstance(xrange, tuple) and len(xrange) == 2):
          xrange = (0,1)
    h = 0; lbl = ''
    if action == 'f':
        h = f
        lbl = 'f'
    elif action == 'df/dx':
        h = f.derivative(x)
        lbl = '\\frac{df}{dx}'
    elif action == 'int f':
        h = f.integrate(x)
        lbl = '\\int f dx'
    elif action == 'num f':
        h = f.numerator()
        lbl = '\\text{numer(f)}'
    elif action == 'den f':
        h = f.denominator()
        lbl = '\\text{denom(f)}'
    elif action == '1/f':
        h = 1/f
        lbl = '\\frac{1}{f}'
    elif action == 'finv':
        h = solve(f == var('y'), x)[0].rhs()
        lbl = 'f^{-1}(y)'
    elif action == 'f+a':
        h = f+a
        lbl = 'f + a'
    elif action == 'f-a':
        h = f-a
        lbl = 'f - a'
    elif action == 'f*a':
        h = f*a
        lbl = 'f \\times a'
    elif action == 'f/a':
        h = f/a
        lbl = '\\frac{f}{a}'
    elif action == 'f^a':
        h = f^a
        lbl = 'f^a'
    elif action == 'f^a':
        h = f^a
        lbl = 'f^a'
    elif action == 'f(x+a)':
        h = f(x+a)
        lbl = 'f(x+a)'
    elif action == 'f(x*a)':
        h = f(x*a)
        lbl = 'f(xa)'
    elif action == 'f+g':
        h = f+g
        lbl = 'f + g'
    elif action == 'f-g':
        h = f-g
        lbl = 'f - g'
    elif action == 'f*g':
        h = f*g
        lbl = 'f \\times g'
    elif action == 'f/g':
        h = f/g
        lbl = '\\frac{f}{g}'
    elif action == 'f(g)':
        h = f(g)
        lbl = 'f(g)'
    pretty_print(html('<center><font color="red">$f = %s$</font></center>'%latex(f)))
    pretty_print(html('<center><font color="green">$g = %s$</font></center>'%latex(g)))
    pretty_print(html('<center><font color="blue"><b>$h = %s = %s$</b></font></center>'%(lbl, latex(h))))
    if do_plot:
        P = plot(f, xrange, color='red', thickness=2) +  \
            plot(g, xrange, color='green', thickness=2) + \
            plot(h, xrange, color='blue', thickness=2)
        if yrange == 'auto':
            show(P, xmin=xrange[0], xmax=xrange[1])
        else:
            yrange = sage_eval(yrange)
            show(P, xmin=xrange[0], xmax=xrange[1], ymin=yrange[0], ymax=yrange[1])
```

1. **Funkcia bez symetrického grafu**:  
   $$f(x) = x^2 + e^x$$
   Táto funkcia je ani párna, ani nepárna, pretože  
   - $ f(-x) = (-x)^2 + e^{-x} = x^2 + e^{-x} $  
   - čo sa nerovná ani $ f(x) $(čo by znamenalo párnosť), ani $ -f(x) $ (čo by znamenalo nepárnosť).

   $$ g(x) = x^3 - 3x + 1$$
   Táto funkcia je tiež ani párna, ani nepárna, pretože:  
   - $ g(-x) = (-x)^3 - 3(-x) + 1 = -x^3 + 3x + 1 $,  
   - čo sa nerovná $ g(x) $ ani $ -g(x) $.  

   **Symetria tejto funkcie**:  
   Graf tejto funkcie je symetrický podľa bodu $[0,1]$, pretože posunutie $g(x) - 1 $ (t.j. $ x^3 - 3x $) je nepárna funkcia.

**8. Pomôžte si Definíciou 8 i grafom funkcie $y=\dfrac{1}{x}$ a zdôvodnite, prečo táto funkcia nie je monotónna na množine $(-\infty,0) \cup (0,\infty)$ (nájdite dve dvojice rôznych bodov a ukážte, že Definícia 8 nie je splnená).**

<ins> Riešenie: <ins>  
    
- **Na intervale $ (-\infty, 0) $**: Funkcia je **klesajúca**. Ak zoberieme dva body $ x_1 < x_2 $ na tomto intervale, platí, že $ f(x_1) > f(x_2) $. Napríklad, keď zoberieme $ x_1 = -2 $ a $ x_2 = -1 $, máme $ f(-2) = -\frac{1}{2} $ a $ f(-1) = -1 $, čo ukazuje, že $ f(-2) > f(-1) $, teda funkcia je klesajúca.

- **Na intervale $ (0, \infty) $**: Funkcia je **klesajúca**. Ak zoberieme dva body $ x_1 < x_2 $ na tomto intervale, platí, že $ f(x_1) > f(x_2) $. Napríklad, keď zoberieme $ x_1 = 1$ a $ x_2 = 2 $, máme $ f(1) = 1 $ a $ f(2) = \frac{1}{2} $, čo ukazuje, že $ f(1) > f(2)$, teda funkcia je klesajúca.
    
**Kontrapríklad:**
Zoberme dve hodnoty $x_1 = -1$ a $x_2 = 1$ z množiny $(-\infty, 0) \cup (0, \infty)$.
- Pre $x_1 = -1$ máme $f(-1) = -1$.
- Pre $x_2 = 1$ máme $f(1) = 1$.

Tu vidíme, že:
- $x_1 < x_2$ (pretože $-1 < 1$),
- ale $f(x_1) = -1 < f(x_2) = 1$.  
    
Toto poruší podmienku monotónnosti, pretože funkcia nie je konzistentne klesajúca ani rastúca na celej množine $(-\infty, 0) \cup (0, \infty)$.

**Záver:**
Aj keď je funkcia klesajúca na intervaloch $(-\infty, 0)$ a $(0, \infty)$ zvlášť, ale na celej množine $(-\infty, 0) \cup (0, \infty)$ nie je monotónna kvôli zmene znamienka funkčných hodnôt pri prechode z negatívnych hodnôt ($x < 0$) na pozitívne ($x > 0$).

#### 1. Graf funkcie môže mať viacero priesečníkov s horizontálnou priamkou (rovnobežka s osou $x$).

<font color=blue> a) Áno </font>   
b) Nie
 
  
<ins> Vysvetlenie: </ins>    
Znamená to, že existuje viacero hodnôt $x$, pre ktoré hodnota funkcie $f(x)$ je rovnaká. Ak si predstavíme horizontálnu priamku $y=c$, priesečník tejto priamky s grafom funkcie $f(x)$ nastane v bodoch, kde $f(x)=c$. Ak je funkcia napríklad periodická alebo má viacero lokálnych extrémov, môže sa stať, že bude mať viacero takýchto priesečníkov s horizontálnou priamkou.

#### 2.  Priraďte predpis funkcie ku grafu:


a) $y=x-5$;  <font color=blue> (V) </font>

b) $-3x+4=y$; <font color=blue> (IV) </font> 

c) $5=y$; <font color=blue> (I) </font>

d) $y=-4x-5$; <font color=blue> (VI) </font>

e) $y=x+6$; <font color=blue> (II) </font> 

f) $y=\dfrac{x}{2}$.  <font color=blue> (III) </font>




<ins> Vysvetlenie: </ins>     
**(I)** Z obrázka vidíme, že graf funkcie je rovnobežný s osou $x$-ovou. Graf funkcie pretína $y$-ovú os na jej kladnej časti. Všetky body z definičného oboru nadobúdajú jednu kladnú hodnotu. Preto možnosť $y=5$ je správna.  

Bolo potrebné podobne preskúmať jednotlivé možnosti na zistenie správnej odpovedi.

Identifikujeme priesečník s osou $y$ a sklon funkcií.  
A) $y=x−5$:   
Priesečník s osou $y$: keď $x=0$, $y=−5$. To znamená, že graf pretína os $y$ v bode $(0, -5)$.  
Sklon: ak $x$ vzrastie o $1$, $y$ vzrastie o $1$. 

B) $−3x+4=y$:  
Priesečník s osou $y$: keď $x=0$, $y=4$. To znamená, že graf pretína os $y$ v bode $(0, 4)$.  
Sklon: ak $x$ vzrastie o $1$, $y$ klesne o $3$ (graf klesá).  

C) $5=y$:  
Toto je horizontálna priamka v bode $y=5$.  

D) $y=−4x−5$:  
Priesečník s osou $y$: keď $x=0$, $y=−5$. To znamená, že graf pretína os $y$ v bode $(0, -5)$.  
Sklon: ak $x$ vzrastie o $1$, $y$ klesne o $4$ (graf klesá prudko).  

E) $y=x+6$:  
Priesečník s osou $y$: keď $x=0$, $y=6$. To znamená, že graf pretína os $y$ v bode $(0, 6)$.  
Sklon: ak $x$ vzrastie o $1$, $y$ vzrastie o $1$.  

F) $y=x/2$:  
Priesečník s osou $y$: keď $x=0, y=0$. To znamená, že graf pretína os $y$ v bode $(0, 0)$.    
Sklon: ak $x$ vzrastie o $2$, $y$ vzrastie o $1$.  


**(II)** Graf neprechádza bodom $(0, 0)$ a z grafu je zrejmé, že je to priamka s kladným sklonom.  
- Graf má priesečník s osou $y$ v kladnom bode.
- Graf stúpa zľava doprava.

Správnou možnosťou je $y=x+6$, pretože:  
Pretína os $y$ v bode $(0, 6)$.  
Má sklon $1$, čo znamená, že stúpa o $1$ jednotku pre každú jednotku zmeny v $x$.  

**(III)** Graf prechádza bodom $(0, 0)$ a z grafu je zrejmé, že je to priamka s kladným sklonom.  
- Graf má priesečník s osou $y$ v $0$.
- Graf stúpa zľava doprava.

Správnou možnosťou je $y=x/2$, pretože:  
Pretína os $y$ v bode $(0, 0)$.  
Má sklon kladný.  

**(IV)** Graf neprechádza bodom $(0, 0)$ a z grafu je zrejmé, že je to priamka so záporným sklonom.    
- Graf má priesečník s osou $y$ v kladnej časti.
- Graf klesá zľava doprava.

Správnou možnosťou je $y=-3x+4$, pretože:  
Nepretína os $y$ v bode $(0, 0)$.  
Má sklon záporný.  
Pretína $y$-ovú os v jej kladnej časti.  

**(V)** Graf neprechádza bodom $(0, 0)$ a z grafu je zrejmé, že je to priamka so záporným sklonom.  
- Graf má priesečník s osou $y$ v zápornej časti.
- Graf rastie zľava doprava. 

Správnou možnosťou je $y=x-5$ pretože:  
Nepretína os $y$ v bode $(0, 0)$.  
Má sklon kladný.  
Pretína $y$-ovú os v jej zápornej časti.  

**(IV)** Graf neprechádza bodom $(0, 0)$ a z grafu je zrejmé, že je to priamka so záporným sklonom.  
- Graf má priesečník s osou y v zápornej časti.
- Graf klesá zľava doprava.  

Správnou možnosťou je $y=-4x-5$ pretože:  
Nepretína os $y$ v bode $(0, 0)$.  
Má sklon záporný.  
Pretína $y$-ovú os v jej zápornej časti.

#### 6. Nech $\mathcal{D}(f)=\langle a, b\rangle$ a $g(x)=f(x-3)$. Potom $\mathcal{D}(g)$ je:

a) $(a,b)$;  
b) $\langle a,b \rangle$;   
c) $\langle a-3,b-3 \rangle$;   
<font color=blue> d) $\langle a+3,b+3 \rangle$.  </font>   
  
  
<ins> Vysvetlenie: </ins>     
Graf funkcie $f(x−3)$ je grafom funkcie $f(x)$ posunutým o $3$ jednotky **doprava**. Tento posun vzniká, pretože funkcia 
$f(x−3)$ nadobúda rovnakú hodnotu v bode $x+3$, ako funkcia $f(x)$ nadobúda v bode $x$. To znamená, že každá hodnota $y$, ktorú má graf funkcie $f(x)$ pri určitom $x$, sa pri funkcii $f(x−3)$ nachádza o $3$ jednotky doprava.

Ak chcete posunúť uzavretý interval $[a,b]$ doprava, je potrebné pridať $3$ jednotky ku koncovým bodom intervalu $a$ a $b$, čím dostanete nový interval $[a+3,b+3]$. Tento princíp posunu sa vzťahuje na všetky body grafu a intervaly, preto celý graf funkcie $f(x)$ posuniete doprava o $3$ jednotky.

#### 7. Nech $\mathcal{H}(f)=(a,\infty)$ a $g(x)=f(x)-4$. Potom $\mathcal{H}(g)$ je: 

<font color=blue> a) $\left(a-4,\infty\right)$;  </font>   
b) $(a+4,\infty)$;   
c) $(a,\infty)$;   
d) Žiadna z možností.   

  
<ins> Vysvetlenie: </ins>   
Ak je obor hodnôt funkcie $f$ daný intervalom $(a,\infty)$, môžeme zapísať nasledujúcu nerovnosť: $f(x)>a$. Pridáme $-4$ na obe strany nerovnosti a získame $f(x)−4>a−4$. Posledná nerovnosť naznačuje, že obor hodnôt funkcie $f(x)−4$ je $(a−4,\infty)$.

#### 10.  Akú periódu má funkcia $g(x)=\cos(x)+2$?
 
a) $\pi$   
<font color=blue> b) $2\pi$   </font>    
c)  $\frac{1}{2}\pi$  
d) $4\pi$ 
  
<ins> Vysvetlenie: </ins>  
Z **Poznámky 10** vieme, že funkcia $\cos(x)$ je periodická s periódou $2\pi$. To znamená, že po každom intervale dĺžky $2\pi$ sa hodnoty funkcie opakujú, teda $\cos(x)=\cos(x+2\pi)$.


Funkcia $g(x)=\cos(x)+2$ vznikne pridaním konštanty $+2$, čo spôsobí, že celý graf funkcie $\cos(x)$ sa posunie o $2$ jednotky nahor na osi $y$.
**Dôležité:** Tento vertikálny posun nemení periodicitu funkcie. Hodnoty funkcie $g(x)$ sa opakujú pri rovnakej perióde ako $cos(x)$, teda $2\pi$.

Napr. zoberieme si bod $x=0$, tak $\cos(0)=1$, ale $\cos(0)+2=3$  
$x=2\pi$, tak $\cos(2\pi)=1$, ale $\cos(2\pi)+2=3$

Vidíme, že iné funkčné hodnoty, ale tie hodnoty sa opakujú tiež pri perióde $2\pi$.

#### 11.  Funkcie $f(t)=5\sin(t)$, $g(t)=8+5\sin(t)$ majú rovnaké amplitúdy.
 
<font color=blue> a) Áno  </font>    
b)  Nie
  
<ins> Vysvetlenie: </ins>  
Amplitúda funkcie sínusu je definovaná ako polovica rozdielu medzi maximálnou a minimálnou hodnotou funkcie. V prípade oboch funkcií $f(t)$ a $g(t)$ je amplitúda určená koeficientom pred $\sin(t)$.

Pre funkciu $f(t) = 5\sin(t)$:  
- Amplitúda je priamo 5, pretože $\sin(t)$ má amplitúdu 1, a je násobená 5. 

Pre funkciu $g(t) = 8 + 5\sin(t)$:   
- Člen 8 je len vertikálny posun celej funkcie a nemá vplyv na amplitúdu.  
- Amplitúda je opäť určená koeficientom 5 pred $\sin(t)$.

Pre lepšie porozumenie doporučujem vykresliť si obidva grafy do jedného grafu a následne porovnať a overiť si svoje odhady.

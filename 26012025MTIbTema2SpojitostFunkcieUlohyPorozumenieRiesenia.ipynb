#### 1. Ktorá z uvedených funkcií je spojitou funkciou času?

a) Počet študentov zapísaných do študijnej skupiny počas semestera.  
<font color=blue> b) Množstvo plynu v nádrži automobilu na ceste medzi New Yorkom a Bostonom. </font>  
c) Vek najstaršej žijúcej osoby.   
d) Počet zákazníkov vstupujúcich do obchodu počas dňa.
  
<ins> Vysvetlenie: </ins>    
Spojitá funkcia času je taká funkcia, ktorá sa mení plynulo bez náhlych skokov alebo prerušení.  

Množstvo plynu v nádrži sa mení postupne a plynulo počas jazdy. Spotreba paliva je kontinuálny proces - ak máme natankovanú plnú nádrž, objem paliva v nádrži v jednotlivých časových okamihoch postupne nadobúda všetky možné hodnoty od maximálneho objemu nádrže po nulu, aj keď merať vieme iba s určitou presnosťou.  

Ostatné možnosti predstavujú diskrétne zmeny alebo potenciálne nespojité funkcie:  

a) Počet študentov sa mení v diskrétnych krokoch (celých číslach) a môže sa meniť náhle - napríklad keď sa študent odhlási alebo prihlási do skupiny, zmena nastane okamžite.  

c) Vek najstaršej žijúcej osoby sa mení skokovo v rokoch (keď niekto zomrie a iná osoba sa stane najstaršou) a nie je spojitou funkciou času.  

d) Počet zákazníkov vstupujúcich do obchodu sa mení v diskrétnych jednotkách (iba po prirodzených číslach, nemôžeme mať polovicu či desatinu zákazníka) a môže mať náhle zmeny - napríklad keď naraz vstúpi do obchodu väčšia skupina ľudí.

#### 2. Ktoré z uvedených funkcií nie sú spojité v bode $x=c$

  

a) iba (I)  
b) iba (II)  
<font color=blue> c) iba (III)</font>    
d) (I) a (II)    
e) (I) a (III)  
f) (II) a (III)  
g) (I), (II) a (III)  
  
<ins> Vysvetlenie: </ins>  
Analyzujme spojitosť každej funkcie v bode $x=c$:  

**Funkcia (I)**:   
Graf je spojitý v bode c.  
Hoci má v bode c "špic", funkcia je v tomto bode spojitá, pretože:  
- Funkcia je v bode c definovaná  
- Existujú jednostranné limity  
- Obe jednostranné limity sa rovnajú funkčnej hodnote v bode c  

**Funkcia (II)**:    
Graf je spojitý v bode c.  
Prechádza bodom c plynulo bez skoku alebo prerušenia.
- Všetky podmienky spojitosti sú splnené.  

**Funkcia (III)**:  
V bode c má funkcia **nespojitosť druhého druhu (neodstrániteľná nespojitosť)**  
- Jednostranné limity existujú, ale sú rôzne
- Limita sprava sa blíži k $+\infty$
- Limita zľava sa blíži k $-\infty$  
Funkcia **nie je** spojitá v bode c

#### 3. Pre ľubovoľnú funkciu $f(x)$ takú, že $f(a)=2$, $f(b)=4$, Darbouxova veta hovorí, že funkcia $f$ nadobúda hodnotu $3$ v nejakom bode $x$ medzi $a$, $b$. Je tento výrok pravdivý?

a) Áno  
<font color=blue> b) Nie </font>

Kontrapríklad:

- V bode a má hodnotu 2
- V bode b má hodnotu 4
- Medzi bodmi a a b má skok (nespojitosť)
- **Preskakuje hodnotu 3**

Darbouxova veta (**Veta 9**) platí len pre **spojité funkcie**.

#### 4. Ak funkcia nie je spojitá v bode, potom nie je definovaná v danom bode. Je tento výrok pravdivý?

a) Áno  
<font color=blue> b) Nie </font>   
  
<ins> Vysvetlenie: </ins>   
Funkcia môže byť definovaná v bode, aj keď v ňom nie je spojitá.   

Existujú príklady funkcií, ktoré sú definované v bode, ale nie sú v ňom spojité:
1. **Funkcia s odstrániteľnou nespojitosťou** - funkcia je definovaná v bode, ale jej hodnota sa nerovná limite.
2. **Funkcia so skokom** - funkcia je definovaná, ale limity zľava a sprava sú rôzne.

Preto samotná definovanosť funkcie v bode nezaručuje jej spojitosť v tomto bode, a naopak, nespojitosť funkcie v bode neznamená, že funkcia nie je v tomto bode definovaná.


Jednoduchý príklad:
Uvažujme funkciu:
$$
f(x) =
\begin{cases}
x^2 & \text{pre } x \neq 2 \\
5 & \text{pre } x = 2
\end{cases}
$$

Táto funkcia je definovaná v bode ( $x = 2$ ) (hodnota je 5), ale nie je v ňom spojitá, pretože:

$$\lim_{x \to 2} f(x) = 4 \neq f(2) = 5$$

Teda funkcia má v bode ( $x = 2$ ) **odstrániteľnú nespojitosť** - je tam definovaná, ale jej hodnota "vyskakuje" z plynulého priebehu funkcie.

#### 5. Súčet spojitých funkcií je spojitá funkcia. Je to pravda?

<font color=blue> a) Áno </font>  
b) Nie 
  
<ins> Vysvetlenie: </ins>  
Ak máme dve spojité funkcie f(x) a g(x), ich súčet $h(x) = f(x) + g(x)$ je tiež **spojitá funkcia**.  

Toto platí pretože:  
1. Limita súčtu sa rovná súčtu limít.
2. Ak existujú limity oboch funkcií v danom bode, bude existovať aj limita ich súčtu.
3. Ak sú obe funkcie spojité v bode x₀, potom:
$$\lim\limits_{x\to x_0} [f(x) + g(x)] = \lim\limits_{x\to x_0} f(x) + \lim\limits_{x\to x_0} g(x) = f(x_0) + g(x_0)$$

Toto pravidlo sa dá rozšíriť aj na súčet konečného počtu spojitých funkcií.

#### 6. Ak je funkcia $f$ spojitá v bode $c$, potom $\lim\limits_{x\to c^+} f(x)=f(c)$. Je to Pravda?

<font color=blue> a) Áno </font>  
b) Nie   
  
<ins> Vysvetlenie: </ins>  
Ak je funkcia f spojitá v bode $c$, potom platí, že **pravostranná limita** sa rovná **funkčnej hodnote** v tomto bode. Toto vyplýva z **Definície 1** spojitosti funkcie v bode aj z **Vety 1**.

#### 7. Ak je funkcia $f$ spojitá na intervale $\langle 0, 1)$ a tiež na intervale $\langle 1, 2)$, potom je $f$ spojitá aj na intervale $\langle 0, 2)$.

a) Áno  
<font color=blue> b) Nie  </font>  
  
<ins> Vysvetlenie: </ins>   
Spojitosť funkcie na dvoch susedných polouzavretých intervaloch nezaručuje spojitosť na ich zjednotení.  
Dôvody sú nasledujúce:  
1. Spojitosť na intervale $\langle 0, 1)$  podľa definície zaručuje spojitosť vo všetkých vnútorných bodoch tohto intervalu a v bode $0$ spojitosť sprava (v bode $1$ vôbec nie je zaručená spojitosť).  
2. Spojitosť na intervale $\langle 1, 2)$ podľa definície zaručuje spojitosť vo všetkých vnútorných bodoch tohto intervalu a v bode $1$ spojitosť sprava (v bode $2$ vôbec nie je zaručená spojitosť).   

3. V bode x = 1, kde sa intervaly spájajú, môže existovať nespojitosť, pretože:
    - Prvý interval neobsahuje bod 1
    - Druhý interval obsahuje bod 1, ale nezaručuje spojitosť zľava
    - Môže existovať skok v bode 1 
    
  
Príklad takejto funkcie (načrtnite si graf):  
$$f(x) = x \space  \text{pre}  \space   x ∈ \langle 0, 1) $$
$$f(x) = x + 1  \space  \text{pre}  \space    x ∈ \langle 1, 2) $$ 
Táto funkcia je spojitá na každom z uvedených intervalov, ale má skok v bode $x = 1$.

#### 8. Existujú dve funkcie $f$, $g$ také, že platí $\lim\limits_{x\to x_0} f(x)=\lim\limits_{x\to x_0} g(x)=f(x_0)$, no funkcia $g$ nie je spojitá v bode $x_0$.

<font color=blue> a) Áno </font>  
b) Nie 
  
<ins> Vysvetlenie: </ins>  
Príklad:  
Nech $f(x)$ je **spojitá funkcia**, napríklad $f(x) = x$  
Funkciu $g(x)$ môžeme definovať ako:  
$g(x) = f(x)$ pre všetky $x \ne x₀$  
$g(x₀) = k$, kde $k \ne f(x₀)$

V tomto prípade:  
$$\lim\limits_{x\to x_0} f(x) = f(x_0)$$ (keďže $f$ je spojitá)
$$\lim\limits_{x\to x_0} g(x) = f(x_0)$$ (keďže $g$ sa líši od $f$ len v jednom bode).  

Ale $$g(x₀) \ne \lim\limits_{x\to x_0} g(x),$$ takže $g$ **nie je spojitá** v $x₀$.

Teda funkcia $g$ má rovnakú limitu ako $f$ v bode x₀, ale **nie je v tomto bode spojitá**, pretože jej **funkčná hodnota sa nezhoduje s limitou**.

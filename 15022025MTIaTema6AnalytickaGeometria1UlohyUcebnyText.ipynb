**2. Na konkrétnych príkladoch ilustrujte platnosť vzťahov uvedených v Tvrdení 1 (tzn. zvoľte si ľubovoľné tri vektory $\mathbf{u}$, $\mathbf{v}$, $\mathbf{w}$, konštantu $c$ a rozpíšte pravú i ľavú stranu vo vzťahoch (i)-(iv)).**

**Ilustrácia platnosti vzťahov z Tvrdenia 1:**

Zvoľme si vektory a konštantu:
- $\mathbf{u} = \begin{pmatrix}1\\2\end{pmatrix}$

- $\mathbf{w} = \begin{pmatrix}5\\6\end{pmatrix}$

**Tvrdenie 1 (i):**  
$\mathbf{u}^T \cdot \mathbf{v} = \mathbf{v}^T \cdot \mathbf{u}$

$\mathbf{v}^T \cdot \mathbf{u} =(3,4)\cdot \begin{pmatrix}1\\2\end{pmatrix}=3 \cdot 1 + 4 \cdot 2 = 3 + 8 = 11$

**Záver:** $11 = 11$  

---

**Tvrdenie 1 (ii):**  
$\mathbf{u}^T \cdot (\mathbf{v} + \mathbf{w}) = \mathbf{u}^T \cdot \mathbf{v} + \mathbf{u}^T \cdot \mathbf{w}$

**Ľavá strana:**  
$\mathbf{v} + \mathbf{w} = \begin{pmatrix}8\\10\end{pmatrix}$

**Záver:** $28 = 28$  

---

**Tvrdenie 1 (iii):**  
$(c\mathbf{u})^T \cdot \mathbf{v} = c(\mathbf{u}^T \cdot \mathbf{v})$

$c\mathbf{u} = 2\cdot \begin{pmatrix}1\\2\end{pmatrix}=\begin{pmatrix}2\\4\end{pmatrix}$

**Tvrdenie 1 (iv):**  
$\mathbf{u}^T \cdot \mathbf{u} \geq 0$ a $\mathbf{u}^T \cdot \mathbf{u} = \mathbf{0}$ práve vtedy, keď $\mathbf{u} = \mathbf{0}$

**Záver:** Platí pre všetky vektory.

Toto dokazuje, že pre nenulový vektor $\mathbf{v}$ je jeho norma nenulová.
2. Potom ukážeme priamu implikáciu:
$||\mathbf{0}|| = \sqrt{0^2 + 0^2} = 0$
Toto dokazuje, že pre nulový vektor je jeho norma nulová.

Kombináciou týchto dvoch krokov sme dokázali ekvivalenciu tvrdenia v oboch smeroch. Použitie obmenenej implikácie v prvom kroku je kľúčové pre úplnosť dôkazu.

**3. Na konkrétnom príklade dvoch vektorov ilustrujte nerovnosť z Vety 1.**

<ins> Riešenie: <ins>  
**Ilustrácia Cauchy-Schwarzovej nerovnosti s konkrétnymi vektormi**

Zvoľme si dvojrozmerné vektory:
- $\mathbf{u} = \begin{pmatrix}3\\4\end{pmatrix}$  
- $\mathbf{v} = \begin{pmatrix}1\\2\end{pmatrix}$  

**a) Vypočítajme ľavú stranu nerovnosti (absolútnu hodnotu skalárneho súčinu):**  
$\mathbf{u}^T \cdot \mathbf{v} = 3 \cdot 1 + 4 \cdot 2 = 3 + 8 = 11$  
$|\mathbf{u}^T \cdot \mathbf{v}| = |11| = 11$  

**b) Vypočítajme pravú stranu nerovnosti (súčin noriem):**  
$||\mathbf{u}|| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5$  
$||\mathbf{v}|| = \sqrt{1^2 + 2^2} = \sqrt{1 + 4} = \sqrt{5} \approx 2.236$  
$||\mathbf{u}|| \cdot ||\mathbf{v}|| = 5 \cdot \sqrt{5} \approx 11.180$  

**c) Porovnanie:**  
$11 \leq 11.180$  

**Záver:**  
Nerovnosť $|\mathbf{u}^T \cdot \mathbf{v}| \leq ||\mathbf{u}|| \cdot ||\mathbf{v}||$ platí.  
V tomto prípade je skalárny súčin menší ako súčin dĺžok vektorov, pretože vektory nie sú rovnobežné.


**Znázornenie v Sage:**  

```python
u = vector(RR, [3, 4])
v = vector(RR, [1, 2])

# Skalárny súčin
dot_product = u * v
abs_dot = abs(dot_product)

# Normy
norm_u = u.norm()
norm_v = v.norm()
product_norms = norm_u * norm_v

show(dot_product)
show(abs_dot)
show(f"||u|| = {float(norm_u):.3f}, ||v|| = {float(norm_v):.3f}")
show(f"||u||⋅||v|| = {float(product_norms):.3f}")
show(f"Platí |{abs_dot}| ≤ {float(product_norms):.3f}")
```

**Výstup:** 
Skalárny súčin: $11$   

$|\mathbf{u}\cdot \mathbf{v}| = 11$  
$||\mathbf{u}|| = 5.000, ||\mathbf{v}|| = 2.236$  
$||\mathbf{u}||⋅||\mathbf{v}|| = 11.180$   
Platí $|11| ≤ 11.180 $

**Vysvetlenie:**  
- Cauchy-Schwarzova nerovnosť **kvantifikuje maximálnu možnú zhodu medzi smermi vektorov**.   
- Rovnosť nastane len pre rovnobežné vektory (jeden je násobkom druhého).
- V príklade vektory nie sú rovnobežné, preto je ľavá strana **ostro menšia** ako pravá strana.

**5. Určte odchýlku (nie kosínus odchýlky) vektorov $\mathbf{u}=(2,1,-2)^T$, $\mathbf{v}=(1,1,1)$.**  


<ins> Riešenie <ins>  
    
Odchýlka medzi vektormi $\mathbf{u}=(2,1,-2)^T$ a $\mathbf{v}=(1,1,1)$ je približne $78.9^\circ$ alebo $1.377$ radiánov.

**Výpočet:**

1. **Skalárny súčin:**
   $$\mathbf{u} \cdot \mathbf{v} = (2)(1) + (1)(1) + (-2)(1) = 1 $$

2. **Dĺžka vektora $\mathbf{u}$:**
   $$\|\mathbf{u}\| = \sqrt{2^2 + 1^2 + (-2)^2} = \sqrt{9} = 3 $$

3. **Dĺžka vektora $\mathbf{v}$:**
   $$\|\mathbf{v}\| = \sqrt{1^2 + 1^2 + 1^2} = \sqrt{3}  $$

4. **Kosínus uhla:**
   $$ \cos\theta = \frac{\mathbf{u} \cdot \mathbf{v}}{\|\mathbf{u}\| \|\mathbf{v}\|} = \frac{1}{3\sqrt{3}} \approx 0.19245 $$

5. **Uhol $\theta$:**
  $$\theta = \arccos\left(\frac{1}{3\sqrt{3}}\right) \approx 78.9^\circ \quad $$



```python
# Vektory
u = np.array([2, 1, -2])
v = np.array([1, 1, 1])

# Skalárny súčin
dot_product = np.dot(u, v)

# Dĺžky vektorov
u_norm = np.linalg.norm(u)
v_norm = np.linalg.norm(v)

# Uhol v radiánoch a stupňoch
cos_theta = dot_product / (u_norm * v_norm)
theta_radians = np.arccos(cos_theta)
theta_degrees = np.degrees(theta_radians)

# Zobrazenie výsledkov
display(dot_product, u_norm, v_norm, cos_theta, theta_radians, theta_degrees)
```

**7. Vypočítajte vektorový súčin vektorov $\mathbf{a}=(0,1,1)^T$, $\mathbf{b}=(3,-1,2)^T$. Označme $\mathbf{c}=\mathbf{a}\times\mathbf{b}$. Overte, že vektor $\mathbf{c}$ je kolmý na $\mathbf{a}$ aj $\boldsymbol{b}$.**

<ins> Riešenie <ins>  

Vypočítajme vektorový súčin vektorov $\mathbf{a}=(0,1,1)^T$ a $\mathbf{b}=(3,-1,2)^T$:

$\mathbf{c} = \mathbf{a} \times \mathbf{b} = \begin{vmatrix} 
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
0 & 1 & 1 \\
3 & -1 & 2
\end{vmatrix}$

$= (1\cdot2 - 1\cdot(-1))\mathbf{i} - (0\cdot2 - 1\cdot3)\mathbf{j} + (0\cdot(-1) - 1\cdot3)\mathbf{k}$

$= 3\mathbf{i} + 3\mathbf{j} - 3\mathbf{k}$

Teda $\mathbf{c} = (3,3,-3)^T$.

Teraz overíme, či je vektor $\mathbf{c}$ kolmý na $\mathbf{a}$ aj $\mathbf{b}$ pomocou skalárneho súčinu:

1) $\mathbf{a} \cdot \mathbf{c} = 0 \cdot 3 + 1 \cdot 3 + 1 \cdot (-3) = 0$

2) $\mathbf{b} \cdot \mathbf{c} = 3 \cdot 3 + (-1) \cdot 3 + 2 \cdot (-3) = 0$

Keďže oba skalárne súčiny sú rovné nule, vektor $\mathbf{c}$ je kolmý na $\mathbf{a}$ aj $\mathbf{b}$.

```python
# Definovanie vektorov a a b
a = vector([0, 1, 1])
b = vector([3, -1, 2])

# Vektorový súčin
c = a.cross_product(b)

# Výpočet skalárneho súčinu
dot_product_1 = a.dot_product(c)
dot_product_2 = b.dot_product(c)

# Výsledky
c, dot_product_1, dot_product_2
```

**8. Zvoľte si tri ľubovoľné konkrétne vektory $\mathbf{a},\mathbf{b},\mathbf{c}$ v 3D, jednu reálnu konštantu $p$ a ilustrujte platnosť vlastností v Tvrdení 4.**

**Ilustrácia platnosti vlastností z Tvrdenia 4:**

Zvoľme si vektory a konštantu:
- $\mathbf{a} = (1,0,0)^T$  
- $\mathbf{b} = (0,1,0)^T$  
- $\mathbf{c} = (0,0,1)^T$  
- $p = 2$

---

**Tvrdenie 4 (i):**  
$||\mathbf{a} \times \mathbf{b}|| = ||\mathbf{a}|| \cdot ||\mathbf{b}|| \cdot \sin\theta$

**Výpočet:**  
$\mathbf{a} \times \mathbf{b} = \begin{vmatrix}\mathbf{i} & \mathbf{j} & \mathbf{k} \\ 1 & 0 & 0 \\ 0 & 1 & 0\end{vmatrix} = \begin{pmatrix}0\\0\\1\end{pmatrix}$  
$||\mathbf{a} \times \mathbf{b}|| = \sqrt{0^2 + 0^2 + 1^2} = 1$  
$||\mathbf{a}|| = 1, \ ||\mathbf{b}|| = 1, \ \theta = 90^\circ \Rightarrow \sin\theta = 1$  
$||\mathbf{a}|| \cdot ||\mathbf{b}|| \cdot \sin\theta = 1 \cdot 1 \cdot 1 = 1$  
**Záver:** platí $1 = 1$   

---

**Tvrdenie 4 (ii):**  
$\mathbf{a} \times \mathbf{b} = -\mathbf{b} \times \mathbf{a}$

**Výpočet:**  
$\mathbf{a} \times \mathbf{b} = \begin{pmatrix}0\\0\\1\end{pmatrix}$  
$\mathbf{b} \times \mathbf{a} = \begin{vmatrix}\mathbf{i} & \mathbf{j} & \mathbf{k} \\ 0 & 1 & 0 \\ 1 & 0 & 0\end{vmatrix} = \begin{pmatrix}0\\0\\-1\end{pmatrix}$  
$-\mathbf{b} \times \mathbf{a} = \begin{pmatrix}0\\0\\1\end{pmatrix}$  
**Záver:** platí $\begin{pmatrix}0\\0\\1\end{pmatrix} = \begin{pmatrix}0\\0\\1\end{pmatrix}$   

---

**Tvrdenie 4 (iii):**  
$(2\mathbf{a}) \times \mathbf{b} = 2(\mathbf{a} \times \mathbf{b}) = \mathbf{a} \times (2\mathbf{b})$

**Výpočet:**  
$2\mathbf{a} = \begin{pmatrix}2\\0\\0\end{pmatrix}$  
$(2\mathbf{a}) \times \mathbf{b} = \begin{pmatrix}0\\0\\2\end{pmatrix}$  
$2(\mathbf{a} \times \mathbf{b}) = 2\begin{pmatrix}0\\0\\1\end{pmatrix} = \begin{pmatrix}0\\0\\2\end{pmatrix}$  
$\mathbf{a} \times (2\mathbf{b}) = \mathbf{a} \times \begin{pmatrix}0\\2\\0\end{pmatrix} = \begin{pmatrix}0\\0\\2\end{pmatrix}$  
**Záver:** Všetky tri výsledky sú $\begin{pmatrix}0\\0\\2\end{pmatrix}$   

---

**Tvrdenie 4 (iv):**  
$\mathbf{a} \times (\mathbf{b} + \mathbf{c}) = \mathbf{a} \times \mathbf{b} + \mathbf{a} \times \mathbf{c}$

**Výpočet:**  
$\mathbf{b} + \mathbf{c} = \begin{pmatrix}0\\1\\1\end{pmatrix}$  
$\mathbf{a} \times (\mathbf{b} + \mathbf{c}) = \begin{vmatrix}\mathbf{i} & \mathbf{j} & \mathbf{k} \\ 1 & 0 & 0 \\ 0 & 1 & 1\end{vmatrix} = \begin{pmatrix}0\\-1\\1\end{pmatrix}$  

$\mathbf{a} \times \mathbf{b} = \begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
1 & 0 & 0 \\
0 & 1 & 0
\end{vmatrix}  = \begin{pmatrix} 0 \\ 0 \\ 1 \end{pmatrix}$

$\mathbf{a} \times \mathbf{c} = \begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
1 & 0 & 0 \\
0 & 0 & 1
\end{vmatrix} =  \begin{pmatrix} 0 \\ -1 \\ 0 \end{pmatrix}$

$\mathbf{a} \times \mathbf{b} + \mathbf{a} \times \mathbf{c} = \begin{pmatrix} 0 \\ 0 \\ 1 \end{pmatrix} + \begin{pmatrix} 0 \\ -1 \\ 0 \end{pmatrix} = \begin{pmatrix} 0 \\ -1 \\ 1 \end{pmatrix}$  

**Záver:** platí $\begin{pmatrix}0\\-1\\1\end{pmatrix} = \begin{pmatrix}0\\-1\\1\end{pmatrix}$

**Tvrdenie 4 (v):**  
$\mathbf{a}^T \cdot (\mathbf{b} \times \mathbf{c}) = (\mathbf{a} \times \mathbf{b})^T \cdot \mathbf{c}$

**Výpočet:**  
$\mathbf{b} \times \mathbf{c} = \begin{vmatrix}\mathbf{i} & \mathbf{j} & \mathbf{k} \\ 0 & 1 & 0 \\ 0 & 0 & 1\end{vmatrix} = \begin{pmatrix}1\\0\\0\end{pmatrix}$  
$\mathbf{a}^T \cdot (\mathbf{b} \times \mathbf{c}) = \begin{pmatrix}1 & 0 & 0\end{pmatrix} \begin{pmatrix}1\\0\\0\end{pmatrix}=1$  
$(\mathbf{a} \times \mathbf{b})^T \cdot \mathbf{c} = \begin{pmatrix}0 & 0 & 1\end{pmatrix} \begin{pmatrix}0\\0\\1\end{pmatrix} = 1$  
**Záver:** platí $1 = 1$

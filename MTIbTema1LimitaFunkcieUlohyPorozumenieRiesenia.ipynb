#### 1. Graf funkcie $f$ je znázornený na obrázku. Určte $\lim\limits_{x\to3} \left(5f(x)\right)$.

a) 5  
<font color=blue>b) 10   </font>  
c) 15  
d) 20  
e) Limita neexistuje  
  
<ins> Vysvetlenie: </ins>    
Hodnota funkcie $f(3)=3$ nemá vplyv na limitu, pretože limita závisí len od správania funkcie v blízkosti bodu $x=3$, nie od funkčnej hodnoty v tomto bode.   

A z grafu vidíme, že keď sa $x$ blíži k $x=3$ sprava alebo zľava tak, hodnota $f(x)$ je $2$. Preto:
$\lim\limits_{x\to3} f(x) = 2$  

Teraz potrebujeme určiť limit $5f(x)$, aplikujeme **Vetu 4c)** :
$\lim\limits_{x\to3} \left(5f(x)\right) = 5 \cdot \lim\limits_{x\to3} f(x) = 5 \cdot 2 = 10$

#### 2. Predpokladajme, že $\lim\limits_{x\to3} f(x)=7$. Platí $\lim\limits_{x\to3}\left(xf(x)\right)=21$ ?
 
<font color=blue> a) Áno </font>  
b) Nie  

  
<ins> Vysvetlenie: </ins>  
Použijeme vlastnosť **limity súčinu** funkcií. Ak máme dve funkcie f(x) a g(x), pre ktoré existujú limity v bode a, potom platí (**Veta 4d)**):
$$\lim\limits_{x\to a} [f(x) \cdot g(x)] = \lim\limits_{x\to a} f(x) \cdot \lim\limits_{x\to a} g(x)$$

V našom prípade máme:
$$\lim\limits_{x\to3} f(x) = 7$$
$$\lim\limits_{x\to3} x = 3$$ (limita identity)

Teraz aplikujeme vlastnosť limity súčinu:
$$\lim\limits_{x\to3} [x \cdot f(x)] = \lim\limits_{x\to3} x \cdot \lim\limits_{x\to3} f(x) = 3 \cdot 7 = 21$$
Teda platí, že $\lim\limits_{x\to3} (xf(x)) = 21$.

#### 3. Predpokladajme, že $\lim\limits_{x\to3} f(x)=7$ a $\lim\limits_{x\to3} \left(f(x)+g(x)\right)=12$. Platí $\lim\limits_{x\to3} g(x) = 5$ ? 

<font color=blue> a) Áno </font>  
b) Nie  
  
<ins> Vysvetlenie: </ins>   
Použijeme vlastnosť **limity súčtu** funkcií. Ak máme dve funkcie f(x) a g(x), pre ktoré existujú limity v bode a, potom platí (**Veta 4c)**):
$$\lim\limits_{x\to a} [f(x) + g(x)] = \lim\limits_{x\to a} f(x) + \lim\limits_{x\to a} g(x)$$

V našom prípade máme:
$$\lim\limits_{x\to3} f(x) = 7$$
$$\lim\limits_{x\to3} [f(x) + g(x)] = 12$$

Dosadíme tieto hodnoty do vlastnosti limity súčtu:
$$\lim\limits_{x\to3} [f(x) + g(x)] = \lim\limits_{x\to3} f(x) + \lim\limits_{x\to3} g(x)$$
$$12 = 7 + \lim\limits_{x\to3} g(x)$$

Odčítaním 7 od oboch strán dostaneme:
$$\lim\limits_{x\to3} g(x) = 12 - 7 = 5$$

Teda platí, že $\lim\limits_{x\to3} g(x) = 5$.

#### 4. Grafy funkcií $f$ a $h$ sú znázornené na obrázku. Určte $\lim\limits_{x\to0} f\left(h(x)\right)$.

a) 4  
b) 5   
c) 1  
d) 0    
<font color=blue> e) Limita neexistuje </font> 
  


<ins> Vysvetlenie: </ins>    

**Najprv určíme správanie $ h(x) $, keď $ x \to 0 $:**  
Z grafu funkcie $ h(x) $ vidíme, že:  
- Keď $ x \to 0^- $ (zľava), $ h(x) \to 3 $.  
- Keď $ x \to 0^+ $ (sprava), $ h(x) \to 1 $.  

Jednostranné limity funkcie $ h(x) $ v bode $ x = 0 $ sú rôzne, takže:  
$$
\lim\limits_{x \to 0} h(x) \text{ neexistuje}.
$$ 

**Preto nemá zmysel analyzovať limitu zloženej funkcie $ f(h(x)) $:**  
Keďže limita vnútornej funkcie $ h(x) $ neexistuje, ani limita zloženej funkcie $ f(h(x)) $ nemôže existovať.

#### 5. Predpokladajme, že $\lim\limits_{x\to7+} f(x)=2$ a $\lim\limits_{x\to7-} f(x)=-2$. Je nasledujúci výrok pravdivý? 
$\lim\limits_{x\to7} f(x)$ existuje.

a) Áno  
<font color=blue> b) Nie </font>  
 
  
<ins> Vysvetlenie: </ins>   
Aby limita funkcie v bode existovala, musia sa zhodovať jednostranné limity zľava a sprava (**Veta 1**). V tomto prípade máme:
$$\lim\limits_{x\to7+} f(x) = 2$$
$$\lim\limits_{x\to7-} f(x) = -2$$

Keďže tieto jednostranné limity sa nerovnajú $(2 ≠ -2)$, limita funkcie f(x) v bode 7 neexistuje.

#### 6. Predpokladajme, že $\lim\limits_{x\to7+} f(x)=2$ a $\lim\limits_{x\to7-} f(x)=-2$. Je nasledujúci výrok pravdivý? 
$\lim\limits_{x\to7} \left(f(x)\right)^2$ existuje.

<font color=blue> a) Áno </font>  
b) Nie  
 
  
<ins> Vysvetlenie: </ins>  
Aj keď limita funkcie f(x) v bode 7 **neexistuje** (keďže jednostranné limity sa nerovnajú), limita **(f(x))^2** v bode 7 **existuje**.  

Dôvod je nasledovný (**Veta 4f)**) :
$$\lim\limits_{x\to7+} \left(f(x)\right)^2 = (\lim\limits_{x\to7+} f(x))^2 = 2^2 = 4$$
$$\lim\limits_{x\to7-} \left(f(x)\right)^2 = (\lim\limits_{x\to7-} f(x))^2 = (-2)^2 = 4$$

Tvrdenia, o limitách, ktoré sú sformulované pre limity ako také sú zväčša platné aj pre jednostranné limity. Keďže obe jednostranné limity $(f(x))^2$ sa rovnajú 4, limita $(f(x))^2$ v bode 7 existuje a je rovná 4.  

Tento príklad ukazuje, že aj keď limita funkcie neexistuje, limita jej druhej mocniny môže existovať. Je to preto, že druhá mocnina "vymaže" znamienko a zjednotí hodnoty oboch jednostranných limít.

#### 7. Nech platí $0\le f(x) \le a(x)$ a $\lim\limits_{x\to0} a(x)=0$. Je nasledujúci výrok pravdivý? 
$\lim\limits_{x\to0} f(x)=0$.

<font color=blue> a) Áno </font>  
b) Nie   
  
<ins> Vysvetlenie: </ins>   
Tento výrok je pravdivý a môžeme ho dokázať pomocou **Vety 6 o zovretí**  

$$\lim\limits_{x\to0} 0 \le \lim\limits_{x\to0} f(x) \le \lim\limits_{x\to0} a(x)$$
$$0 \le \lim\limits_{x\to0} f(x) \le 0$$
Z toho vyplýva, že $\lim\limits_{x\to0} f(x) = 0$.

#### 8. Ak existuje vlastná limita $\lim\limits_{x\to0}\dfrac{f(x)}{g(x)}$, tak existujú tiež vlastné limity $\lim\limits_{x\to0} f(x)$ a $\lim\limits_{x\to0} g(x)$.

a) Áno   
<font color=blue> b) Nie </font>  
 
  
<ins> Vysvetlenie: </ins>   
Existencia vlastnej limity podielu dvoch funkcií nezaručuje existenciu vlastných limít jednotlivých funkcií. Môže nastať situácia, kedy limity $\lim\limits_{x\to0} f(x)$ a $\lim\limits_{x\to0} g(x)$ neexistujú alebo sú nevlastné, ale ich podiel má vlastnú limitu.  

Príklad:  
Uvažujme funkcie $f(x) = \frac{1}{x}$ a $g(x) = \frac{1}{x^2}$.  

Pre tieto funkcie platí:  
$$\lim\limits_{x\to0} f(x) = \lim\limits_{x\to0} \frac{1}{x}$$ je nevlastná limita (diverguje k nekonečnu).
$$\lim\limits_{x\to0} g(x) = \lim\limits_{x\to0} \frac{1}{x^2}$$ je tiež nevlastná limita (diverguje k nekonečnu rýchlejšie ako $f(x)$).  

Ale ich podiel má vlastnú limitu:  
$$\lim\limits_{x\to0} \frac{f(x)}{g(x)} = \lim\limits_{x\to0} \frac{\frac{1}{x}}{\frac{1}{x^2}} = \lim\limits_{x\to0} x = 0$$  

Tento príklad ukazuje, že môže existovať vlastná limita podielu dvoch funkcií, aj keď limity oboch funkcií sú nevlastné.

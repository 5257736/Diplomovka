#### 1. $y=x^2$ je jediná taká funkcia, ktorej derivácia je $y'=2x$.

a) Áno  
<font color=blue> b) Nie </font> 
  
<ins> Vysvetlenie: </ins>    
Všetky funkcie, ktoré majú deriváciu $2x$, môžeme zapísať ako:  
$$y = x^2 + C$$
kde C je ľubovoľná konštanta. 

Dôvody:
- Derivácia konštanty je 0, takže pridanie konštanty C k funkcii $x^2$ nezmení jej deriváciu.
- Napríklad tieto funkcie majú všetky deriváciu $2x$:
$$y = x^2$$
$$y = x^2 + 1$$
$$y = x^2 - 5$$
$$y = x^2 + \pi$$
- Všetky takéto funkcie sa líšia len o konštantu a ich grafy dostaneme posunutím grafu funkcie $y = x^2$ vo vertikálnom smere.

Preto existuje **nekonečne veľa** funkcií, ktorých derivácia je $2x$.

#### 2. Ak $f(x)=\dfrac{1}{x^2}$, potom $f'(x)=\dfrac{1}{2x}$.

a) Áno  
<font color=blue> b) Nie </font> 
  
<ins> Vysvetlenie: </ins>  
Pre funkciu $f(x)=\frac{1}{x^2}$ je jej derivácia:  
$$f'(x) = -\frac{2}{x^3}$$

1. spôsob (**derivácia mocninovej funkcie**)  

    Toto môžeme overiť pomocou pravidiel pre derivovanie:  
    $f(x)=\frac{1}{x^2}$ môžeme zapísať ako $f(x)=x^{-2}$

    Pri derivovaní mocninovej funkcie $x^n$ násobíme exponentom a znížime mocninu o 1. Teda $f'(x)=-2x^{-3}=-\frac{2}{x^3}$.



2. spôsob (**derivácia podielu funkcií**)
$$f(x) = \frac{g(x)}{h(x)} = \frac{1}{x^2} = \frac{1}{x} \cdot \frac{1}{x}$$

    kde:
    - $ g(x) = 1 $ (konštantná funkcia),
    - $ h(x) = x^2 $.

    Teraz použijeme pravidlo derivácie podielu:

    $$f'(x) = \frac{g'(x) h(x) - g(x) h'(x)}{[h(x)]^2}$$

     Derivácia $ g(x) = 1 $ je $ g'(x) = 0 $,
     Derivácia $ h(x) = x^2 $ je $ h'(x) = 2x $.

    Teraz dosadíme tieto hodnoty do vzorca:

$$f'(x) = \frac{0 \cdot x^2 - 1 \cdot 2x}{(x^2)^2}$$

$$f'(x) = \frac{-2x}{x^4}$$

$$f'(x) = -\frac{2}{x^3}$$

Výsledok je rovnaký ako pri použití pravidla pre deriváciu mocninovej funkcie. Je jedno, ktorý spôsob si zvolíme.

#### 3. Neexistuje funkcia $f$ taká, pre ktorú by platilo $f'(x)=f(x)$ pre všetky $x$.

a) Áno  
<font color=blue> b) Nie </font>  

<ins> Vysvetlenie: </ins>  
Existuje funkcia, pre ktorú platí $f'(x)=f(x)$ pre všetky $x$.  
Je to exponenciálna funkcia:  
$$f(x) = e^x$$ 

Pre túto funkciu platí:  
        a) $f(x) = e^x$  
        b) $f'(x) = e^x$  

Teda $f'(x) = f(x)$ pre všetky $x$.
Exponenciálna funkcia $e^x$ je jedinečná v tom, že jej derivácia je rovná samotnej funkcii.

#### 4. Funkcia $f(x)$ a jej derivácia $f'(x)$ majú rovnaký definičný obor.

a) Áno  
<font color=blue> b) Nie </font> 
  
<ins> Vysvetlenie: </ins>    
Funkcia môže byť definovaná pre všetky $x$, ale nemusí byť **diferencovateľná (nemať deriváciu)** v každom bode.  
Existuje niekoľko dôvodov:  
- Funkcia môže mať "špic" **(ostrý bod)** - napríklad funkcia $f(x) = |x|$ je definovaná pre všetky $x$, ale nie je diferencovateľná v bode $x = 0$.
- Funkcia môže mať **skok** - napríklad funkcia:
$$f(x) = 1\space \text{pre} \space x ≥ 0$$
$$f(x) = 0 \space \text{pre} \space x < 0$$
    je definovaná pre všetky $x$, ale nie je diferencovateľná v bode $x = 0$ (**Veta 2**).
- Funkcia môže mať **nespojitosť** - napríklad funkcia $f(x) = x^{1/2}$ je definovaná pre všetky $x \geq 0$, ale nie je diferencovateľná v bode $x = 0$.   

Preto samotná definovanosť funkcie pre všetky $x$ **nezaručuje existenciu derivácie pre všetky $x$**.

#### 5. Nikdy nenastane rovnosť $(f\cdot g)'(x)=f'(x)\cdot g'(x)$.

a) Áno       
<font color=blue> b) Nie   </font> 
  
<ins> Vysvetlenie: </ins>   
Podľa **Veta 3** platí, že $$(f\cdot g)'(x_0) = f'(x_0)\cdot g(x_0) + f(x_0)\cdot g'(x_0)$$  

Táto rovnosť však môže nastať napríklad pre konštantné funkcie. Ak zoberieme:  

$f(x) = c$ (konštantná funkcia)  
$g(x) = d$ (konštantná funkcia)  

Potom:  
$(f\cdot g)'(x) = 0$ (derivácia konštanty je 0)  
$f'(x) = 0$ (derivácia konštanty)  
$g'(x) = 0$ (derivácia konštanty)  

A teda: $(f\cdot g)'(x) = 0 = 0\cdot 0 = f'(x)\cdot g'(x)$

#### 6. Derivácia funkcie $f(x)=\ln(\ln\,x)$ je $f'(x)=\dfrac{1}{x}\cdot\ln\,x+\ln\,x\cdot\dfrac{1}{x}=\dfrac{2\ln\,x}{x}$.

a) Áno  
<font color=blue> b) Nie </font>  
  
<ins> Vysvetlenie: </ins>  
Derivácia funkcie $f(x)=\ln(\ln,x)$ sa vypočíta pomocou pravidla o zloženej funkcii (**Veta 4**) (reťazové pravidlo):  
- Vonkajšia funkcia je $\ln(u)$ s deriváciou $\frac{1}{u}$
- Vnútorná funkcia je $u=\ln(x)$ s deriváciou $\frac{1}{x}$

Aplikáciou reťazového pravidla dostávame:  
$$f'(x) = \frac{1}{\ln(x)} \cdot \frac{1}{x} = \frac{1}{x\ln(x)}$$

Tvrdenie v zadaní uvádza nesprávny výsledok $f'(x)=\frac{2\ln,x}{x}$, ktorý sa líši od správneho výsledku $f'(x)=\frac{1}{x\ln(x)}$.

#### 7. Derivácia funkcie $f(x)$ v bode $x_0$ je dotyčnica ku grafu funkcie $f(x)$ v bode $x_0$. Je to pravda?

a) Áno  
<font color=blue> b) Nie </font>  
  
<ins> Vysvetlenie: </ins>   
Derivácia funkcie $f(x)$ v bode $x_0$ nie je dotyčnica, ale je to číslo, ktoré vyjadruje smernicu (sklon) dotyčnice ku grafu funkcie v danom bode.

Toto číslo predstavuje:
- Smernicu dotyčnice v danom bode
- Okamžitú zmenu funkcie v danom bode
- Rýchlosť zmeny funkčnej hodnoty v danom bode

#### 8. Derivácia lineárej funkcie je konštantná funkcia. Platí to?
​
<font color=blue> a) Áno  </font>   
b) Nie  
  
<ins> Vysvetlenie: </ins>   
Toto môžeme dokázať nasledovne:
- Lineárna funkcia má tvar f(x) = ax + b, kde a,b sú konštanty
- Použitím základného vzorca pre deriváciu polynómov vieme, že:  
    člen $ax$ sa derivuje na $a$ (koeficient $a$ krát derivácia $x^1$, podľa vzorca vieme, že znižujeme exponent o 1, $x^{1-1}=x^0=1$)
    konštanta $b$ sa derivuje na 0 (derivácia konštanty je 0)

Teda: $f'(x) = a\cdot 1+0=a$

Výsledkom je konštanta $a$, ktorá je smernicou pôvodnej lineárnej funkcie.  
Preto je derivácia lineárnej funkcie **vždy konštantná funkcia, nezávislá od $x$.**

#### 9. Ak $g(x)$ je vertikálne posunutá funkcia $f(x)$, tzn. $g(x)=f(x)+c$; $c\in\mathbb{R}$, potom $f'(x)=g'(x)$. Je to pravda?
​
<font color=blue> a) Áno  </font>   
b) Nie  
  
<ins> Vysvetlenie: </ins>    
Ak je $g(x)$ vertikálne posunutá funkcia $f(x)$ o konštantu $c$, teda $g(x) = f(x) + c$, tak ich derivácie sa rovnajú, lebo pri derivovaní súčtu funkcie a konštanty je derivácia konštanty $0$.
$$g'(x) = (f(x) + c)' = f'(x) + 0 = f'(x)$$

#### 10. Na ktorom z uvedených intervalov nespĺňa funkcia $y=1/x$ predpoklady Lagrangeovej vety o strednej hodnote?

a) $\langle-3,-1\rangle$;

<font color=blue> b) $\langle-1,1\rangle$; </font>     

c) $\langle2,8\rangle$.

<ins> Vysvetlenie: </ins>   
**Lagrangeova veta o strednej hodnote (Veta 6)** vyžaduje, aby funkcia spĺňala tieto predpoklady:  
1. Funkcia musí byť spojitá na uzavretom intervale $\langle a,b\rangle$.
2. Funkcia musí byť diferencovateľná na otvorenom intervale $(a,b)$.

Funkcia $y = \frac{1}{x}$:  
- Nie je definovaná v bode $x = 0$
- Má nespojitosť v bode $x = 0$
- Interval $\langle-1,1\rangle$ obsahuje bod $x = 0$

Preto na intervale $\langle-1,1\rangle$ funkcia nespĺňa predpoklady Lagrangeovej vety, pretože **nie je spojitá** na celom intervale a má nespojitosť **v bode 0**, ktorý leží vnútri intervalu. Podľa **Vety 2** o spojitosti a derivovateľnosti vieme, že ak funkcia nie je spojitá v bode, nemôže v ňom byť ani diferencovateľná. Teda v bode $x = 0$ neexistuje derivácia funkcie.

a) Nula riešení.  
<font color=blue> b) Aspoň jedno riešenie. </font>  
 
  
<ins> Vysvetlenie: </ins>    
Vektor $\mathbf{b}$ sa dá vyjadriť ako lineárna kombinácia stĺpcov matice $\mathsf{A}$ vtedy a len vtedy, ak existuje aspoň jeden vektor $\mathbf{x}$, ktorý, keď ho prenásobíme maticou $\mathsf{A}$, dostaneme výsledok rovný $\mathbf{b}$. Teda, ak existuje nejaký vektor $\mathbf{x}$, pre ktorý platí $\mathsf{A}\mathbf{x}=\mathbf{b}$, potom vektor $\mathbf{b}$ je lineárnou kombináciou stĺpcov matice $\mathsf{A}$. Naopak, ak neexistuje žiadne riešenie tejto rovnice, vektor $\mathbf{b}$ nie je možné vyjadriť ako lineárnu kombináciu stĺpcov matice $\mathsf{A}$.

#### 3. Koľko riešení môže mať sústava lineárnych rovníc nad poľom $\mathbb{R}$ ?

<font color=blue> a) Žiadne.  </font>  
<font color=blue> b) Jedno.    </font>  
c) Aspoň dve, ale konečne veľa.   
<font color=blue> d) Nekonečne veľa.   </font>
  
<ins> Vysvetlenie: </ins>   
Počet riešení sústavy lineárnych rovníc nad poľom $\mathbb{R}$ môže byť:

**Jedno riešenie**: Ak sústava má jedno riešenie, to znamená, že existuje práve jedna hodnota pre každú neznámu (premennú), pričom po dosadení týchto hodnôt do sústavy rovníc, všetky rovnice sú splnené súčasne (ľavá strana sa rovná pravej strane).

**Nekonečne veľa riešení**: Ak sústava má nekonečne veľa riešení, to sa stáva vtedy, keď sústava obsahuje nejaké voľné premenné, tzn. viac premenných ako rovníc ("nadbytočne" premenne sa stávajú, parametrami, za ktoré dosadzujeme rôzne reálne čísla, čím dostávame nekonečne veľa rôznych riešení.

**Žiadne riešenie**: Ak sústava lineárnych rovníc nemá žiadne riešenie, to znamená, že nie je možné nájsť žiadne hodnoty pre neznáme, ktoré by spĺňali všetky rovnice sústavy súčasne.

Počet riešení sústavy závisí od počtu rovníc, neznámych a ich vzájomných vzťahov. Môže sa stať, že sústava bude mať jedno riešenie, nekonečne veľa riešení alebo žiadne riešenie.

#### 4. Koľko (rôznych - nekongruentných) riešení môže mať sústava lineárnych rovníc nad poľom $\mathbb{Z}_5$ ?

<font color=blue>  a) Žiadne.   </font>  
<font color=blue>   b) Konečne veľa.  </font>     
c) Nekonečne veľa.  

  
<ins> Vysvetlenie: </ins>    
Máme k dispozícii 5 rôznych prvkov (zvyškových tried), ktoré môžeme dosadzať za neznáme (premenné), čiže takto vieme nakombinovať iba konečne veľa rôznych "n-tíc" resp. konečne veľa rôznych riešení. 

**Ilustračný príklad 1**  
Nad poľom  $\mathbb{Z}_5$ (konečné pole so zvyškami po delení 5, teda s prvkami ${0,1,2,3,4}$), môžeme zvoliť sústavu dvoch rovníc s dvomi neznámymi $x$ a $y$ takto:

$$   2x+3y≡1\mod 5$$
$$   4x+y≡3\mod 5$$

Na riešenie tejto sústavy budeme hľadať hodnoty $x$ a $y$, ktoré spĺňajú obe rovnice $\mod5$.

Riešenie:  
**Krok 1**: Vyjadrenie jednej neznámej z druhej rovnice
Z druhej rovnice vyjadríme $y$:

$$y=3−4x $$
Preložíme do kladných čísel a zjednodušíme:

$$y≡3−4x≡3+x $$
**Krok 2**: Dosadenie do prvej rovnice
Dosadíme $y=3+x$ do prvej rovnice:

$$2x+3(3+x)=1 $$

$$2x+9+3x=1  $$

$$5x+9=1 $$

$$0+4=1 $$
Táto sústava **nemá riešenie** nad poľom  $\mathbb{Z}_5$.  

**Ilustračný príklad 2**
$$   x+2y≡3\mod 5$$
$$   3x+4y≡2\mod 5$$

Na riešenie tejto sústavy budeme hľadať hodnoty $x$ a $y$, ktoré spĺňajú obe rovnice $\mod5$.

Riešenie:  
**Krok 1**: Z prvej rovnice môžeme vyjadriť $x$:

$$x=3-2y $$

**Krok 2**: Dosadíme $x=3-2y$ do druhej rovnice:

$$3(3−2y)+4y=2$$

$$9−6y+4y=2$$

$$9−2y=2$$

$$4−2y=2$$

$$−2y=−2$$

$$y=1$$

**Krok 3**:
Keď $y=1$, dosadíme do prvej rovnice:

$$x+2(1)=3$$

$$x+2=3$$

$$x=1$$
Riešením sústavy je $x=1$ a $y=1$ nad poľom  $\mathbb{Z}_5$.

#### 5. Každá homogénna sústava lineárnych rovníc nad $\mathbb{R}$ má riešenie.

<font color=blue> a) Áno   </font>  
 b) Nie     
  
<ins> Vysvetlenie: </ins>     
Každá homogénna sústava lineárnych rovníc nad množinou reálnych čísel $\mathbb{R}$ má vždy aspoň jedno riešenie, a to tzv. **triviálne riešenie**, kedy všetky neznáme sú rovné nule.  

Toto vyplýva z vlastností lineárnych rovníc a z **definície homogénnej sústavy**. Homogénna sústava je taká, kde všetky pravé strany rovníc sú nulové. Preto je vždy možné nájsť riešenie, keď všetky neznáme sú nulové, čo spĺňa podmienky homogénnej sústavy.  

Samozrejme, môže existovať aj ďalšie **ne-triviálne** riešenia, v závislosti od počtu rovníc a neznámych, ale existencia triviálneho riešenia je zaručená.   


**Ilustračný príklad**
$\mathsf{A}\mathbf{x}=0$, $\mathbf{x}$ je vektor, $\mathsf{A}$ je matica  
Ak dáme za $\mathbf{x}$ nulový vektor tak, po vynásobení vektora $\mathbf{x}$ zľava ľubovoľnou maticou $\mathsf{A}$ sústavy, to je $\mathsf{A}\mathbf{x}$, dostaneme nulový vektor, čo je pravá strana sústavy a teda $x=0$, čo znamená, že je to **triviálne riešenie**.

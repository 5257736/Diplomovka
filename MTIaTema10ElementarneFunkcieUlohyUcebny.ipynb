Vykreslite niekoľko rôznych grafov v Sage a svoje pozorovania popíšte slovne.**

<ins> Riešenie: <ins>

Na experimentovanie s lineárnou funkciou v tvare $y = ax + b$ použijeme Sage na vykreslenie niekoľkých grafov s rôznymi hodnotami koeficientov $a$ a $b$. Pozorujme, ako sa mení graf a vlastnosti funkcie pri zmene týchto koeficientov.

Začnime so základnou lineárnou funkciou:
- Tento graf predstavuje funkciu $y = x$, kde $a = 1$ a $b = 0$. Je to priamka prechádzajúca počiatkom so sklonom 45°.

```python
def f(x, a, b):
    return a*x + b

x = var('x')
plot(f(x, 1, 0), (x, -5, 5), ymin=-5, ymax=5, title='y = x')
```

Zmena koeficientu $a$
- Teraz zmeníme koeficient $a$

- Pozorovania:
    1. Keď $a > 1$ (modrá čiara), sklon priamky sa zvyšuje a funkcia rastie rýchlejšie.
    2. Keď $0 < a < 1$ (červená čiara), sklon priamky sa znižuje a funkcia rastie pomalšie.
    3. Keď $a < 0$ (zelená čiara), priamka má záporný sklon a funkcia klesá.

```python
p1 = plot(f(x, 2, 0), (x, -5, 5), color='blue', legend_label='y = 2x')
p2 = plot(f(x, 0.5, 0), (x, -5, 5), color='red', legend_label='y = 0.5x')
p3 = plot(f(x, -1, 0), (x, -5, 5), color='green', legend_label='y = -x')
show(p1 + p2 + p3, ymin=-5, ymax=5, title='Zmena koeficientu a')
```

Zmena koeficientu $b$
- Teraz zmeníme koeficient $b$
- Pozorovania:
    1. Keď $b > 0$ (modrá čiara), graf sa posúva nahor o $b$ jednotiek.
    2. Keď $b < 0$ (červená čiara), graf sa posúva nadol o $|b|$ jednotiek.
    3. Keď $b = 0$ (zelená čiara), graf prechádza počiatkom.

```python
p1 = plot(f(x, 1, 2), (x, -5, 5), color='blue', legend_label='y = x + 2')
p2 = plot(f(x, 1, -2), (x, -5, 5), color='red', legend_label='y = x - 2')
p3 = plot(f(x, 1, 0), (x, -5, 5), color='green', legend_label='y = x')
show(p1 + p2 + p3, ymin=-5, ymax=5, title='Zmena koeficientu b')
```

Kombinácia zmien $a$ a $b$
- Nakoniec skombinujeme zmeny oboch koeficientov
- Pozorovania:
    1. Modrá čiara ($a > 1, b > 0$): strmý rast, posunutá nahor.
    2. Červená čiara ($a < 0, b < 0$): mierny pokles, posunutá nadol.
    3. Zelená čiara ($a = 0, b \neq 0$): horizontálna priamka, konštantná funkcia.

```python
p1 = plot(f(x, 2, 1), (x, -5, 5), color='blue', legend_label='y = 2x + 1')
p2 = plot(f(x, -0.5, -1), (x, -5, 5), color='red', legend_label='y = -0.5x - 1')
p3 = plot(f(x, 0, 3), (x, -5, 5), color='green', legend_label='y = 3')
show(p1 + p2 + p3, ymin=-5, ymax=5, title='Kombinácia zmien a a b')
```

Experimentovaním s lineárnou funkciou $y = ax + b$ v Sage sme pozorovali:
    
   1. Koeficient $a$ určuje sklon priamky:
        - Kladné $a$: funkcia rastie
        - Záporné $a$: funkcia klesá
        - Väčšie $|a|$: strmší sklon
        - Menšie $|a|$: miernejší sklon


   2. Koeficient $b$ určuje posun grafu po osi y:
       - Kladné $b$: posun nahor
       - Záporné $b$: posun nadol
       - $b = 0$: graf prechádza počiatkom


   3. Kombinácia $a$ a $b$ umožňuje vytvoriť širokú škálu lineárnych funkcií s rôznymi vlastnosťami.

Tieto experimenty názorne ukazujú, ako jednoduché zmeny koeficientov môžu výrazne ovplyvniť správanie lineárnej funkcie.

Vykreslite niekoľko rôznych grafov v Sage a svoje pozorovania popíšte slovne.**

<ins> Riešenie: <ins>

Na experimentovanie s kvadratickou funkciou v tvare $y = ax^2 + bx + c$ použijeme Sage na vykreslenie niekoľkých grafov s rôznymi hodnotami koeficientov $a$, $b$ a $c$. Pozorujme, ako sa mení graf a vlastnosti funkcie pri zmene týchto koeficientov.

Začnime so základnou kvadratickou funkciou:
- Tento graf predstavuje funkciu $y = x^2$, kde $a = 1$, $b = 0$ a $c = 0$. Je to parabola s vrcholom v počiatku, otvorená nahor.

```python
def f(x, a, b, c):
    return a*x^2 + b*x + c

x = var('x')
plot(f(x, 1, 0, 0), (x, -5, 5), ymin=-5, ymax=5, title='y = x^2')
```

Zmena koeficientu $a$

- Teraz zmeníme koeficient $a$
- Pozorovania:
    1. Keď $a > 1$ (modrá krivka), parabola je užšia a strmšia.
    2. Keď $0 < a < 1$ (červená krivka), parabola je širšia a plytšia.
    3. Keď $a < 0$ (zelená krivka), parabola je otvorená nadol.

```python
p1 = plot(f(x, 2, 0, 0), (x, -5, 5), color='blue', legend_label='y = 2x^2')
p2 = plot(f(x, 0.5, 0, 0), (x, -5, 5), color='red', legend_label='y = 0.5x^2')
p3 = plot(f(x, -1, 0, 0), (x, -5, 5), color='green', legend_label='y = -x^2')
show(p1 + p2 + p3, ymin=-5, ymax=5, title='Zmena koeficientu a')
```

Zmena koeficientu $b$

- Teraz zmeníme koeficient $b$
- Pozorovania:
    1. Keď $b \neq 0$, os symetrie paraboly sa posúva horizontálne.
    2. Keď $b > 0$ (modrá krivka), parabola sa posúva doľava.
    3. Keď $b < 0$ (červená krivka), parabola sa posúva doprava.

```python
p1 = plot(f(x, 1, 2, 0), (x, -5, 5), color='blue', legend_label='y = x^2 + 2x')
p2 = plot(f(x, 1, -2, 0), (x, -5, 5), color='red', legend_label='y = x^2 - 2x')
p3 = plot(f(x, 1, 0, 0), (x, -5, 5), color='green', legend_label='y = x^2')
show(p1 + p2 + p3, ymin=-5, ymax=5, title='Zmena koeficientu b')
```

Zmena koeficientu $c$
- Nakoniec zmeníme koeficient $c$
- Pozorovania:
    1. Keď $c > 0$ (modrá krivka), celá parabola sa posúva nahor.
    2. Keď $c < 0$ (červená krivka), celá parabola sa posúva nadol.
    3. Koeficient $c$ nemení tvar paraboly, len jej vertikálnu polohu.

```python
p1 = plot(f(x, 1, 0, 2), (x, -5, 5), color='blue', legend_label='y = x^2 + 2')
p2 = plot(f(x, 1, 0, -2), (x, -5, 5), color='red', legend_label='y = x^2 - 2')
p3 = plot(f(x, 1, 0, 0), (x, -5, 5), color='green', legend_label='y = x^2')
show(p1 + p2 + p3, ymin=-5, ymax=5, title='Zmena koeficientu c')
```

Experimentovaním s kvadratickou funkciou $y = ax^2 + bx + c$ v Sage sme pozorovali:

   1. Koeficient $a$ určuje "strmosť" a smer otvorenia paraboly:
       - Kladné $a$: parabola otvorená nahor
       - Záporné $a$: parabola otvorená nadol
       - Väčšie $|a|$: užšia parabola
       - Menšie $|a|$: širšia parabola


   2. Koeficient $b$ ovplyvňuje horizontálny posun vrcholu paraboly:
       - Kladné $b$: posun vrcholu doľava
       - Záporné $b$: posun vrcholu doprava
       - $b = 0$: os symetrie prechádza osou y


   3. Koeficient $c$ určuje vertikálny posun celej paraboly:
       - Kladné $c$: posun nahor
       - Záporné $c$: posun nadol
       - $c = 0$: vrchol leží na osi x (ak $b = 0$)

Tieto experimenty názorne ukazujú, ako zmeny koeficientov ovplyvňujú tvar, polohu a orientáciu grafu kvadratickej funkcie.

**3. Pomocou Sage načrtnite graf funkcie $y=\dfrac{1}{x^2+4}$. Určte, aká priamka je horizontálnou asymptotou grafu tejto funkcie a načrtnite ju.**

<ins> Riešenie: <ins>

Najprv vytvoríme graf danej funkcie:

```python
var('x')
f(x) = 1 / (x^2 + 4)
p = plot(f, (x, -10, 10), ymin=0, ymax=0.3)
show(p)
```

Horizontálna asymptota tejto funkcie je priamka $y = 0$. Môžeme to odvodiť nasledovne:
$$\lim_{x \to \pm\infty} \frac{1}{x^2+4} = 0$$

Keďže pre veľmi veľké hodnoty x sa menovateľ blíži k nekonečnu, celý zlomok sa blíži k nule.

Teraz pridáme horizontálnu asymptotu do nášho grafu:

```python
asymptota = line([(x, 0) for x in [-10, 10]], color='red', linestyle='--')
p_final = p + asymptota
show(p_final)
```

Graf funkcie $y=\frac{1}{x^2+4}$ má tvar zvonu, ktorý je symetrický podľa osi $y$. Funkcia dosahuje maximum v bode $(0, 1/4)$ a postupne klesá k nule pre $x$ blížiace sa k plus alebo mínus nekonečnu.

Horizontálna asymptota grafu je priamka $y = 0$, ktorú sme znázornili červenou prerušovanou čiarou. Graf funkcie sa k tejto asymptote približuje, ale nikdy ju nedosiahne, pretože funkcia je vždy kladná pre všetky reálne hodnoty $x$.

**5. Vykreslite v Sage grafy funkcií $f(t)=(1.029)^t$, $g(t)=14.235\cdot(1.029)^t$ z Príkladu 6. V čom sa líšia grafy? Čo spôsobilo v grafe funkcie $f(t)$ prenásobenie konštantou $14.235$? Čo by sa stalo s grafom funkcie $f(t)$, keď by sme $f(t)$ prenásobili číslom $0.2$ ? Pomôžte si so Sage.**  


<ins> Riešenie: <ins>  
    
Na vykreslenie grafov funkcií $f(t)=(1.029)^t$ a $g(t)=14.235\cdot(1.029)^t$ v Sage použijeme nasledujúci kód:

```python
var('t')
f(t) = (1.029)^t
g(t) = 14.235 * (1.029)^t

p1 = plot(f, (t, 0, 10), color='blue', legend_label='f(t) = (1.029)^t')
p2 = plot(g, (t, 0, 10), color='red', legend_label='g(t) = 14.235 * (1.029)^t')

show(p1 + p2, ymin=0, ymax=20, title='Porovnanie grafov f(t) a g(t)')
```

Rozdiely medzi grafmi:

1. Vertikálne posunutie - Graf funkcie $g(t)$ je vertikálne posunutý nahor v porovnaní s grafom funkcie $f(t)$.

2. Rozsah hodnôt - Graf funkcie $g(t)$ dosahuje výrazne vyššie hodnoty než graf funkcie $f(t)$.

Prenásobenie konštantou $14.235$ spôsobilo:
- Vertikálne roztiahnutie grafu funkcie $f(t)$
- Posun grafu nahor

- Vertikálnemu stlačeniu grafu funkcie $f(t)$
- Posunu grafu nadol smerom k osi x

Na ilustráciu tohto efektu môžeme použiť nasledujúci kód v Sage:

```python
h(t) = 0.2 * (1.029)^t

p3 = plot(h, (t, 0, 10), color='green', legend_label='h(t) = 0.2 * (1.029)^t')

show(p1 + p2 + p3, ymin=0, ymax=20, title='Porovnanie grafov f(t), g(t) a h(t)')
```

Tento graf názorne ukáže, ako prenásobenie funkcie $f(t)$ číslom $0.2$ ovplyvní jej graf - bude stlačený smerom k osi x a bude mať menší rozsah hodnôt v porovnaní s pôvodnou funkciou $f(t)$.

Majú funkcie $y=\log x$, $y=\ln x$ verikálnu asymptotu? Ak áno, je rovnaká pre obe funkcie?**

Na vykreslenie grafov funkcií $y=10^x$, $y=\log x$, $y=x$ a $y=\ln x$ v Sage použijeme nasledujúci kód:

```python
var('x')
p1 = plot(10^x, (x, -2, 2), color='blue', legend_label='y = 10^x')
p2 = plot(log(x, 10), (x, 0.1, 10), color='red', legend_label='y = log x')
p3 = plot(x, (x, -2, 10), color='green', legend_label='y = x')
p4 = plot(ln(x), (x, 0.1, 10), color='purple', legend_label='y = ln x')

show(p1 + p2 + p3 + p4, xmin=-2, xmax=10, ymin=-2, ymax=10, aspect_ratio=1)
```

Pozorovania:

1. Rýchlosť rastu:
   - Funkcia $y=10^x$ rastie exponenciálne a veľmi rýchlo prekračuje ostatné funkcie.
   - Funkcia $y=\log x$ rastie oveľa pomalšie než lineárna funkcia $y=x$.
   - Funkcia $y=\ln x$ rastie mierne pomalšie než $y=\log x$, ale má podobný tvar.


2. Vertikálne asymptoty:
   - Funkcie $y=\log x$ a $y=\ln x$ majú vertikálnu asymptotu v $x = 0$.
   - Táto asymptota je rovnaká pre obe funkcie.


3. Priesečníky:
   - Exponenciálna funkcia $y=10^x$ pretína priamku $y=x$ v bode $(1,1)$.
   - Logaritmické funkcie $y=\log x$ a $y=\ln x$ pretínajú priamku $y=x$ v bode $(1,0)$.


4. Oblasti definície:
   - Exponenciálna funkcia $y=10^x$ je definovaná pre všetky reálne čísla.
   - Logaritmické funkcie $y=\log x$ a $y=\ln x$ sú definované len pre kladné reálne čísla $(x > 0)$.


5. Správanie v nekonečne:
   - Exponenciálna funkcia $y=10^x$ rastie neobmedzene rýchlejšie než ostatné funkcie.
   - Logaritmické funkcie $y=\log x$ a $y=\ln x$ rastú veľmi pomaly a sú "prekonané" lineárnou funkciou $y=x$.


Exponenciálna funkcia $y=10^x$ vykazuje najrýchlejší rast smerom do nekonečna, zatiaľ čo logaritmické funkcie $y=\log x$ a $y=\ln x$ rastú najpomalšie. Obe logaritmické funkcie majú rovnakú vertikálnu asymptotu v $x = 0$.

Viete popísať či a ako sa zmení perióda oproti funkciám $\sin x$, $\cos x$ ? Grafy funkcií môžete kresliť v Sage s využitím ktorýchkoľvek kódov z témy 9 a 10 alebo môžete využiť interaktívny kód na konci učebného textu k téme 10.** 

<ins> Riešenie: <ins>

Na ilustráciu horizontálnych a vertikálnych posunov grafov sínusoidálnych funkcií použijeme Sage. Vykreslíme grafy niekoľkých funkcií a porovnáme ich s základnými funkciami $\sin(x)$ a $\cos(x)$.

```python
var('x')
p1 = plot(sin(x), (x, -2*pi, 2*pi), color='blue', legend_label='sin(x)')
p2 = plot(cos(x), (x, -2*pi, 2*pi), color='yellow', legend_label='cos(x)')
p3 = plot(sin(3*x), (x, -2*pi, 2*pi), color='red', legend_label='sin(3x)')
p4 = plot(2*cos(x), (x, -2*pi, 2*pi), color='green', legend_label='2cos(x)')
p5 = plot(2*cos(4*x), (x, -2*pi, 2*pi), color='purple', legend_label='2cos(4x)')
p6 = plot(4*sin(3*x)-2, (x, -2*pi, 2*pi), color='orange', legend_label='4sin(3x)-2')

combined_plot1 = p1 + p3
combined_plot2 = p2 + p4 
combined_plot3 = p2 + p5
combined_plot4 = p1 + p6
combined_plot = p1 + p2 + p3 + p4 + p5 + p6
combined_plot1.show(ymin=-5, ymax=5)
combined_plot2.show(ymin=-5, ymax=5)
combined_plot3.show(ymin=-5, ymax=5)
combined_plot4.show(ymin=-5, ymax=5)
combined_plot.show(ymin=-5, ymax=5)
```

Pozorovania:
1. $\sin(3x)$ vs. $\sin(x)$:
   - Perióda sa zmenšila 3-krát
   - Amplitúda zostala rovnaká


2. $2\cos(x)$ vs. $\cos(x)$:
   - Perióda zostala rovnaká
   - Amplitúda sa zväčšila 2-krát


3. $2\cos(4x)$ vs. $\cos(x)$:
   - Perióda sa zmenšila 4-krát
   - Amplitúda sa zväčšila 2-krát


4. $4\sin(3x)-2$ vs. $\sin(x)$:
   - Perióda sa zmenšila 3-krát
   - Amplitúda sa zväčšila 4-krát
   - Celý graf je posunutý o 2 jednotky nadol

Všeobecné pravidlá:

1. Horizontálny posun:
   - Funkcia v tvare $\sin(kx)$ alebo $\cos(kx)$, kde k > 1, má periódu zmenšenú k-krát
   - Čím väčšie k, tým menšia perióda a "stlačenejší" graf


2. Vertikálny posun:
   - Násobenie amplitúdy (napr. $2\sin(x)$) zväčšuje výšku vlny, ale nemení periódu
   - Pripočítanie alebo odčítanie konštanty (napr. $\sin(x)+2$) posúva celý graf hore alebo dole, ale nemení periódu ani amplitúdu


3. Kombinácia posunov:
   - Funkcia v tvare $A\sin(kx)+B$  alebo $A\cos(kx)+B$ kombinuje všetky tieto efekty
   - A ovplyvňuje amplitúdu
   - k ovplyvňuje periódu
   - B posúva graf vertikálne

Tieto pozorovania nám pomáhajú pochopiť, ako rôzne transformácie ovplyvňujú správanie sínusoidálnych funkcií. Horizontálne "stlačenie" mení periódu, zatiaľ čo vertikálne transformácie ovplyvňujú amplitúdu a polohu grafu, ale nemenia periódu.

**9. Uvažujte funkciu $g(x)=1/x$. Aj s využitím interaktívneho kódu nájdite predpisy (a vykreslite grafy) pre nasledujúce transformácie grafu tejto funkcie:**

* **(I) posun o 2 hore;** 


* **(II) posun o 3 dole;** 


* **(III) posun o 4 vpravo;** 


* **(IV) posun o 5 vľavo;** 


* **(V) natiahnutie o 2 v smere osi $y$;** 


* **(VI) stiahnutie o 3 v smere osi $x$;** 


* **(VII) prevrátenie "hore nohami".**

<ins> Riešenie: <ins>

Na základe pravidiel transformácií funkcií a s využitím interaktívneho kódu v Sage môžeme pre funkciu  $g(x)=\frac{1}{x}$ vytvoriť nasledujúce transformácie:

 (I) Posun o 2 hore:
       $g(x)=\frac{1}{x}+2$

 (II) Posun o 3 dole:
       $g(x)=\frac{1}{x}-3$

 (III) Posun o 4 vpravo:
       $g(x)=\frac{1}{x-4}$

 (IV) Posun o 5 vľavo:
       $g(x)=\frac{1}{x+5}$

 (V) Natiahnutie o 2 v smere osi y:
       $g(x)=\frac{2}{x}$

 (VI) Stiahnutie o 3 v smere osi x:
       $g(x)=\frac{1}{3\cdot x}$

 (VII) Prevrátenie "hore nohami":
       $g(x)=-\frac{1}{x}$

```python
var('x')
p = plot(1/x , (x, -10, 10), color='black', legend_label='1/x')
p1 = plot(1/x + 2, (x, -10, 10), color='blue', legend_label='1/x + 2')
p2 = plot(1/x - 3, (x, -10, 10), color='yellow', legend_label='1/x - 3')
p3 = plot(1/(x - 4), (x, -10, 10), color='red', legend_label='1/(x - 4)')
p4 = plot(1/(x + 5), (x, -10, 10), color='green', legend_label='1/(x + 5)')
p5 = plot(2/x, (x, -10, 10), color='purple', legend_label='2/x')
p6 = plot(1/(3*x), (x, -10, 10), color='orange', legend_label='1/(3*x)')
p7 = plot(-1/x, (x, -10, 10), color='sienna', legend_label='-1/x')

combined_plot1 = p + p1
combined_plot2 = p + p2 
combined_plot3 = p + p3
combined_plot4 = p + p4
combined_plot5 = p + p5
combined_plot6 = p + p6
combined_plot7 = p + p7
combined_plot = p+p1 + p2 + p3 + p4 + p5 + p6 +p7

combined_plot1.show(ymin=-5, ymax=5)
combined_plot2.show(ymin=-5, ymax=5)
combined_plot3.show(ymin=-5, ymax=5)
combined_plot4.show(ymin=-5, ymax=5)
combined_plot5.show(ymin=-5, ymax=5)
combined_plot6.show(ymin=-5, ymax=5)
combined_plot7.show(ymin=-5, ymax=5)
combined_plot.show(ymin=-5, ymax=5)
```

Pre funkciu $g(x)=\frac{1}{x}$ a jej transformácie je vhodné zvoliť rozsah, ktorý zahŕňa aj záporné, aj kladné hodnoty, ale vynecháva nulu (keďže funkcia nie je v nule definovaná).

Tieto transformácie menia tvar a polohu pôvodného grafu funkcie $g(x)=\frac{1}{x}$ nasledovne:

- Vertikálne posuny ((I) a (II)) posúvajú celý graf nahor alebo nadol.
- Horizontálne posuny ((III) a (IV)) posúvajú asymptotu grafu vpravo alebo vľavo.
- Vertikálne natiahnutie (V) zväčšuje vzdialenosť grafu od osi $x$.
- Horizontálne stiahnutie (VI) zužuje graf smerom k osi $y$.
- Prevrátenie (VII) zrkadlí graf cez os $x$.

Všetky tieto transformácie zachovávajú základný tvar hyperboly, ktorý je charakteristický pre funkciu $g(x)=\frac{1}{x}$, ale menia jej polohu, orientáciu alebo "strmosť".
